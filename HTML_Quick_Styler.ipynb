```python
# -*- coding: utf-8 -*-
"""HTML Quick Styler - v3 — with many additional templates."""

# !pip install -q gradio

import gradio as gr
import base64
import mimetypes
import os
import json
import re
from datetime import datetime
from typing import Tuple, Dict, List, Optional


# ============================================================================
# ColorPaletteGenerator
# ============================================================================
class ColorPaletteGenerator:
    @staticmethod
    def generate_palette(style: str = "modern") -> Dict[str, str]:
        palettes = {
            "modern":   {"primary": "#667eea", "secondary": "#764ba2", "accent": "#f093fb",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "vibrant":  {"primary": "#ff6b6b", "secondary": "#ff8e72", "accent": "#f5576c",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "ocean":    {"primary": "#0099cc", "secondary": "#006699", "accent": "#00ccff",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "forest":   {"primary": "#27ae60", "secondary": "#1e8449", "accent": "#52be80",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "sunset":   {"primary": "#ff8c42", "secondary": "#ff6b35", "accent": "#ffa751",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "luxury":   {"primary": "#1a1a1a", "secondary": "#333333", "accent": "#FFD700",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "creative": {"primary": "#9b59b6", "secondary": "#8e44ad", "accent": "#e74c3c",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
            "minimal":  {"primary": "#2c3e50", "secondary": "#34495e", "accent": "#3498db",
                         "background": "#0f172a", "surface": "#1e293b",
                         "text_primary": "#f1f5f9", "text_secondary": "#cbd5e1", "border": "#334155"},
        }
        return palettes.get(style.lower(), palettes["modern"])


# ============================================================================
# ImageProvider
# ============================================================================
class ImageProvider:
    IMAGES = {
        "hero":        ["1029", "1040", "1042", "1043", "1044"],
        "about":       ["1005", "1006", "1008", "1010", "1012"],
        "portfolio":   ["1015", "1016", "1018", "1020", "1021",
                        "1022", "1023", "1024", "1025", "1026"],
        "team":        ["1027", "1028", "1032", "1033", "1035"],
        "testimonial": ["1036", "1038", "1039", "1041", "1045"],
    }

    @staticmethod
    def picsum(context: str, index: int = 0, width: int = 800, height: int = 500) -> str:
        pool   = ImageProvider.IMAGES.get(context, ImageProvider.IMAGES["portfolio"])
        pic_id = pool[index % len(pool)]
        return f"https://picsum.photos/id/{pic_id}/{width}/{height}"

    @staticmethod
    def avatar(index: int = 0, size: int = 120) -> str:
        pool   = ImageProvider.IMAGES["team"]
        pic_id = pool[index % len(pool)]
        return f"https://picsum.photos/id/{pic_id}/{size}/{size}"

    @staticmethod
    def file_to_data_url(filepath: str) -> str:
        try:
            mime, _ = mimetypes.guess_type(filepath)
            if not mime or not mime.startswith("image/"):
                mime = "image/jpeg"
            with open(filepath, "rb") as f:
                b64 = base64.b64encode(f.read()).decode("utf-8")
            return f"data:{mime};base64,{b64}"
        except Exception:
            return ""

    @staticmethod
    def process_uploads(file_list) -> List[Dict]:
        if not file_list:
            return []
        if isinstance(file_list, str):
            file_list = [file_list]
        results = []
        for item in file_list:
            filepath = item if isinstance(item, str) else getattr(item, "name", str(item))
            if not filepath or not os.path.isfile(filepath):
                continue
            mime, _ = mimetypes.guess_type(filepath)
            if not mime or not mime.startswith("image/"):
                continue
            data_url = ImageProvider.file_to_data_url(filepath)
            if not data_url:
                continue
            size_kb = round(os.path.getsize(filepath) / 1024, 1)
            results.append({
                "name":     os.path.basename(filepath),
                "path":     filepath,
                "data_url": data_url,
                "size_kb":  size_kb,
            })
        return results


# ============================================================================
# HTMLPageBuilder
# ============================================================================
class HTMLPageBuilder:
    @staticmethod
    def extract_sections(description: str) -> List[str]:
        sections = []
        section_keywords = {
            "hero":         ["hero", "banner", "landing"],
            "features":     ["features", "services", "benefits"],
            "about":        ["about", "team", "company"],
            "portfolio":    ["portfolio", "gallery", "work", "projects"],
            "testimonials": ["testimonial", "reviews", "feedback"],
            "pricing":      ["pricing", "plans", "cost"],
            "contact":      ["contact", "form", "reach"],
            "footer":       ["footer", "bottom"],
        }
        desc_lower = description.lower()
        for section_type, keywords in section_keywords.items():
            for kw in keywords:
                if kw in desc_lower:
                    sections.append(section_type)
                    break
        if not sections:
            sections = ["hero", "features", "contact", "footer"]
        return sections

    @staticmethod
    def generate_meta_tags(title: str, description: str) -> str:
        return f"""    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{title}</title>
    <meta name="description" content="{description[:160]}">
    <meta name="theme-color" content="#0f172a">
    <meta property="og:title" content="{title}">
    <meta property="og:description" content="{description[:160]}">
    <meta property="og:type" content="website">
    <meta name="twitter:card" content="summary_large_image">
    <meta name="author" content="HTML Quick-Styler">"""

    @staticmethod
    def generate_section_html(section_type, palette, uploaded_images=None):
        uploaded_images = uploaded_images or []
        if uploaded_images:
            hero_src = uploaded_images[0]["data_url"]
        else:
            hero_src = ImageProvider.picsum("hero", 0, 1400, 700)

        custom_portfolio_cards = ""
        for i, img in enumerate(uploaded_images[1:7]):
            label    = img["name"] or f"Image {i + 2}"
            data_url = img["data_url"]
            fallback = ImageProvider.picsum("portfolio", i, 400, 300)
            custom_portfolio_cards += f"""
                <div class="portfolio-item custom-image-item">
                    <img src="{data_url}" alt="{label}" onerror="this.src='{fallback}'">
                    <div class="portfolio-overlay custom-badge"><span>&#11088; {label}</span></div>
                </div>"""

        sections = {
            "hero": f"""        <section class="hero"
             style="background-image:url('{hero_src}');background-size:cover;background-position:center;">
            <div class="hero-overlay"></div>
            <div class="hero-content">
                <h1>Welcome to Your Site</h1>
                <p>Create amazing web experiences</p>
                <button class="btn btn-primary">Get Started</button>
            </div>
        </section>""",
            "features": f"""        <section class="features">
            <h2>Features</h2>
            <div class="features-grid">
                <div class="feature-card">
                    <img src="{ImageProvider.picsum('portfolio', 0, 60, 60)}" alt="Fast" class="feature-icon">
                    <h3>Fast</h3><p>Lightning-quick performance</p>
                </div>
                <div class="feature-card">
                    <img src="{ImageProvider.picsum('portfolio', 1, 60, 60)}" alt="Secure" class="feature-icon">
                    <h3>Secure</h3><p>Enterprise-grade security</p>
                </div>
                <div class="feature-card">
                    <img src="{ImageProvider.picsum('portfolio', 2, 60, 60)}" alt="Scalable" class="feature-icon">
                    <h3>Scalable</h3><p>Grow without limits</p>
                </div>
            </div>
        </section>""",
            "about": f"""        <section class="about">
            <div class="about-inner">
                <img src="{ImageProvider.picsum('about', 0, 500, 380)}" alt="About Us" class="about-img">
                <div class="about-content">
                    <h2>About Us</h2>
                    <p>We are passionate about creating beautiful, functional websites that help your business grow.</p>
                </div>
            </div>
        </section>""",
            "portfolio": f"""        <section class="portfolio">
            <h2>Our Work</h2>
            <div class="portfolio-grid">{custom_portfolio_cards}
                <div class="portfolio-item">
                    <img src="{ImageProvider.picsum('portfolio', 0, 400, 300)}" alt="Project 1">
                    <div class="portfolio-overlay"><span>Project 1</span></div>
                </div>
                <div class="portfolio-item">
                    <img src="{ImageProvider.picsum('portfolio', 1, 400, 300)}" alt="Project 2">
                    <div class="portfolio-overlay"><span>Project 2</span></div>
                </div>
                <div class="portfolio-item">
                    <img src="{ImageProvider.picsum('portfolio', 2, 400, 300)}" alt="Project 3">
                    <div class="portfolio-overlay"><span>Project 3</span></div>
                </div>
            </div>
        </section>""",
            "testimonials": f"""        <section class="testimonials">
            <h2>What People Say</h2>
            <div class="testimonials-grid">
                <div class="testimonial">
                    <img src="{ImageProvider.avatar(0)}" alt="Client" class="testimonial-avatar">
                    <p>"Amazing service!"</p><span>- Client Name</span>
                </div>
                <div class="testimonial">
                    <img src="{ImageProvider.avatar(1)}" alt="Customer" class="testimonial-avatar">
                    <p>"Highly recommended!"</p><span>- Happy Customer</span>
                </div>
            </div>
        </section>""",
            "pricing": """        <section class="pricing">
            <h2>Pricing Plans</h2>
            <div class="pricing-grid">
                <div class="pricing-card">
                    <h3>Starter</h3><p class="price">$29</p>
                    <ul><li>Feature 1</li><li>Feature 2</li></ul>
                </div>
                <div class="pricing-card featured">
                    <h3>Pro</h3><p class="price">$79</p>
                    <ul><li>All Starter</li><li>Feature 3</li></ul>
                </div>
            </div>
        </section>""",
            "contact": """        <section class="contact">
            <h2>Get in Touch</h2>
            <form class="contact-form">
                <input type="text"  placeholder="Your Name"    required>
                <input type="email" placeholder="Your Email"   required>
                <textarea placeholder="Your Message" rows="5"  required></textarea>
                <button type="submit" class="btn btn-primary">Send Message</button>
            </form>
        </section>""",
            "footer": """        <footer class="footer">
            <div class="footer-content">
                <p>&copy; 2024 Your Company. All rights reserved.</p>
                <div class="footer-links">
                    <a href="#privacy">Privacy</a>
                    <a href="#terms">Terms</a>
                    <a href="#contact">Contact</a>
                </div>
            </div>
        </footer>""",
        }
        return sections.get(section_type, "")

    @staticmethod
    def generate_css(palette: Dict) -> str:
        return f"""    <style>
        * {{ margin:0; padding:0; box-sizing:border-box; }}
        :root {{
            --primary:        {palette['primary']};
            --secondary:      {palette['secondary']};
            --accent:         {palette['accent']};
            --background:     {palette['background']};
            --surface:        {palette['surface']};
            --text-primary:   {palette['text_primary']};
            --text-secondary: {palette['text_secondary']};
            --border:         {palette['border']};
        }}
        body {{
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: linear-gradient(135deg, var(--background) 0%, var(--surface) 100%);
            color: var(--text-primary);
            line-height: 1.6;
            min-height: 100vh;
        }}
        header {{
            background: rgba(15,23,42,0.95);
            backdrop-filter: blur(10px);
            padding: 1.5rem 2rem;
            position: sticky; top:0; z-index:100;
            border-bottom: 1px solid var(--border);
        }}
        header nav {{
            display:flex; justify-content:space-between; align-items:center;
            max-width:1200px; margin:0 auto;
        }}
        header h1 {{
            font-size:1.8rem;
            background: linear-gradient(135deg, var(--primary), var(--accent));
            -webkit-background-clip:text; -webkit-text-fill-color:transparent;
            background-clip:text;
        }}
        header nav a {{
            color:var(--text-secondary); text-decoration:none;
            margin:0 1.5rem; transition:color 0.3s;
        }}
        header nav a:hover {{ color:var(--primary); }}
        main {{ max-width:1200px; margin:0 auto; padding:0 2rem; }}
        section {{ padding:4rem 0; animation:fadeIn 0.6s ease-in; }}
        @keyframes fadeIn {{
            from {{ opacity:0; transform:translateY(20px); }}
            to   {{ opacity:1; transform:translateY(0); }}
        }}
        h1,h2,h3 {{ color:var(--text-primary); margin-bottom:1rem; }}
        h1 {{ font-size:3rem; font-weight:700; }}
        h2 {{ font-size:2rem; margin-bottom:2rem; text-align:center; }}
        h3 {{ font-size:1.3rem; }}
        .hero {{
            position:relative; color:white; padding:6rem 0;
            text-align:center; min-height:500px;
            display:flex; align-items:center; justify-content:center;
            overflow:hidden;
        }}
        .hero-overlay {{
            position:absolute; inset:0;
            background:linear-gradient(135deg,rgba(0,0,0,.6),rgba(0,0,0,.4));
            z-index:1;
        }}
        .hero-content {{ position:relative; z-index:2; }}
        .hero h1 {{ color:white; font-size:4rem; margin-bottom:1rem; animation:slideDown .8s ease; }}
        .hero p  {{ font-size:1.5rem; margin-bottom:2rem; opacity:.95; animation:slideUp .8s ease; }}
        @keyframes slideDown {{ from{{opacity:0;transform:translateY(-30px)}} to{{opacity:1;transform:translateY(0)}} }}
        @keyframes slideUp   {{ from{{opacity:0;transform:translateY(30px)}}  to{{opacity:1;transform:translateY(0)}} }}
        .btn {{
            padding:.8rem 2rem; border:none; border-radius:8px;
            font-size:1rem; cursor:pointer; transition:all .3s; font-weight:600;
        }}
        .btn-primary {{ background:var(--primary); color:white; box-shadow:0 4px 15px rgba(0,0,0,.3); }}
        .btn-primary:hover {{ transform:translateY(-2px); box-shadow:0 6px 20px rgba(0,0,0,.4); background:var(--accent); }}
        .btn-secondary {{ background:var(--secondary); color:white; }}
        .btn-secondary:hover {{ background:var(--primary); }}
        .feature-icon {{
            width:60px; height:60px; border-radius:50%;
            object-fit:cover; margin-bottom:1rem; border:2px solid var(--primary);
        }}
        .features-grid {{
            display:grid; grid-template-columns:repeat(auto-fit,minmax(250px,1fr)); gap:2rem;
        }}
        .feature-card {{
            background:rgba(30,41,59,.8); padding:2rem; border-radius:12px;
            border:1px solid var(--border); transition:all .3s; backdrop-filter:blur(10px);
        }}
        .feature-card:hover {{ transform:translateY(-8px); border-color:var(--primary); box-shadow:0 20px 40px rgba(0,0,0,.3); }}
        .feature-card h3 {{ color:var(--primary); margin-bottom:.5rem; }}
        .feature-card p  {{ color:var(--text-secondary); }}
        .about-inner {{ display:flex; align-items:center; gap:3rem; flex-wrap:wrap; }}
        .about-img {{ width:100%; max-width:460px; border-radius:16px; object-fit:cover; flex:1 1 300px; box-shadow:0 20px 40px rgba(0,0,0,.4); }}
        .about-content {{ flex:1 1 280px; }}
        .about p {{ font-size:1.1rem; color:var(--text-secondary); }}
        .portfolio-grid {{
            display:grid; grid-template-columns:repeat(auto-fit,minmax(280px,1fr)); gap:2rem;
        }}
        .portfolio-item {{
            position:relative; aspect-ratio:4/3; border-radius:12px;
            overflow:hidden; cursor:pointer; border:1px solid var(--border);
        }}
        .custom-image-item {{ border:2px solid var(--accent); box-shadow:0 0 24px rgba(255,255,255,.12); }}
        .portfolio-item img {{
            width:100%; height:100%; object-fit:cover; transition:transform .4s; display:block;
        }}
        .portfolio-overlay {{
            position:absolute; inset:0;
            background:linear-gradient(to top,rgba(0,0,0,.7) 0%,transparent 60%);
            display:flex; align-items:flex-end; padding:1rem;
            opacity:0; transition:opacity .3s;
        }}
        .custom-badge {{ opacity:1 !important; background:linear-gradient(to top,rgba(0,0,0,.8),transparent 80%); }}
        .portfolio-overlay span {{ color:white; font-weight:600; font-size:1rem; }}
        .portfolio-item:hover img {{ transform:scale(1.08); }}
        .portfolio-item:hover .portfolio-overlay {{ opacity:1; }}
        .testimonials-grid {{
            display:grid; grid-template-columns:repeat(auto-fit,minmax(300px,1fr)); gap:2rem;
        }}
        .testimonial {{ background:rgba(30,41,59,.8); padding:2rem; border-left:4px solid var(--primary); border-radius:8px; }}
        .testimonial-avatar {{ width:70px; height:70px; border-radius:50%; object-fit:cover; margin-bottom:1rem; border:3px solid var(--primary); }}
        .testimonial p   {{ font-style:italic; margin-bottom:1rem; color:var(--text-secondary); }}
        .testimonial span{{ font-weight:600; color:var(--primary); }}
        .pricing-grid {{
            display:grid; grid-template-columns:repeat(auto-fit,minmax(300px,1fr)); gap:2rem;
        }}
        .pricing-card {{ background:rgba(30,41,59,.8); padding:2rem; border-radius:12px; border:1px solid var(--border); text-align:center; transition:all .3s; }}
        .pricing-card.featured {{ border-color:var(--primary); box-shadow:0 20px 40px rgba(0,0,0,.3); transform:scale(1.05); }}
        .pricing-card .price {{ font-size:2.5rem; color:var(--primary); margin:1rem 0; font-weight:700; }}
        .pricing-card ul {{ list-style:none; margin:1.5rem 0; text-align:left; }}
        .pricing-card li {{ padding:.5rem 0; color:var(--text-secondary); }}
        .contact {{ max-width:600px; margin:0 auto; }}
        .contact-form {{ display:flex; flex-direction:column; gap:1.5rem; }}
        .contact-form input,
        .contact-form textarea {{
            padding:1rem; border:1px solid var(--border); border-radius:8px;
            background:rgba(30,41,59,.8); color:var(--text-primary);
            font-family:inherit; font-size:1rem;
        }}
        .contact-form input:focus,
        .contact-form textarea:focus {{ outline:none; border-color:var(--primary); box-shadow:0 0 0 3px rgba(0,0,0,.3); }}
        .footer {{ background:rgba(15,23,42,.95); border-top:1px solid var(--border); margin-top:4rem; padding:3rem 0; }}
        .footer-content {{ max-width:1200px; margin:0 auto; padding:0 2rem; display:flex; justify-content:space-between; align-items:center; flex-wrap:wrap; gap:2rem; }}
        .footer-links {{ display:flex; gap:2rem; }}
        .footer-links a {{ color:var(--text-secondary); text-decoration:none; transition:color .3s; }}
        .footer-links a:hover {{ color:var(--primary); }}
        @media(max-width:768px){{
            header nav {{ flex-direction:column; gap:1rem; }}
            header h1  {{ font-size:1.4rem; }}
            .hero h1   {{ font-size:2rem; }}
            .hero p    {{ font-size:1.1rem; }}
            h2         {{ font-size:1.5rem; }}
            main       {{ padding:0 1rem; }}
            section    {{ padding:2rem 0; }}
            .footer-content {{ flex-direction:column; text-align:center; }}
        }}
    </style>"""


# ============================================================================
# HTMLGenerationEngine
# ============================================================================
class HTMLGenerationEngine:
    @staticmethod
    def generate_page(title, description, style="modern", uploaded_images=None):
        palette       = ColorPaletteGenerator.generate_palette(style)
        sections      = HTMLPageBuilder.extract_sections(description)
        meta_tags     = HTMLPageBuilder.generate_meta_tags(title, description)
        css           = HTMLPageBuilder.generate_css(palette)
        sections_html = "\n\n".join([
            HTMLPageBuilder.generate_section_html(s, palette, uploaded_images)
            for s in sections
            if HTMLPageBuilder.generate_section_html(s, palette, uploaded_images)
        ])
        html = f"""<!DOCTYPE html>
<html lang="en">
<head>
{meta_tags}
{css}
</head>
<body>
    <header>
        <nav>
            <h1>{title}</h1>
            <div>
                <a href="#features">Features</a>
                <a href="#about">About</a>
                <a href="#contact">Contact</a>
            </div>
        </nav>
    </header>
    <main>
{sections_html}
    </main>
</body>
</html>"""
        return html, palette


# ============================================================================
# PREMIUM TEMPLATES  (self-contained HTML strings)
# ============================================================================
class PremiumTemplates:
    """Each method returns a complete, self-contained HTML page string."""

    # ── 4. Restaurant / Food ──────────────────────────────────────────────────
    @staticmethod
    def restaurant(title="The Hearth Kitchen"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=Playfair+Display:ital,wght@0,400;0,700;1,400&family=Lato:wght@300;400;700&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
body{{font-family:'Lato',sans-serif;background:#1a0f0a;color:#f0e8d8;line-height:1.8;}}
header{{padding:1.5rem 5rem;background:rgba(26,15,10,.9);backdrop-filter:blur(10px);position:fixed;width:100%;top:0;z-index:99;border-bottom:1px solid rgba(200,140,60,.2);display:flex;justify-content:space-between;align-items:center;}}
.logo{{font-family:'Playfair Display',serif;color:#c88c3c;font-size:1.6rem;}}
nav a{{color:rgba(240,232,216,.65);text-decoration:none;margin-left:2.5rem;font-size:.85rem;letter-spacing:.08em;text-transform:uppercase;transition:color .3s;}}
nav a:hover{{color:#c88c3c;}}
.reserve-btn{{border:1px solid #c88c3c;color:#c88c3c!important;padding:.5rem 1.4rem;}}
.hero{{min-height:100vh;background:linear-gradient(to bottom,rgba(26,15,10,.4) 0%,rgba(26,15,10,.7) 60%,rgba(26,15,10,1) 100%),url('https://picsum.photos/id/1080/1600/900') center/cover;display:flex;flex-direction:column;align-items:center;justify-content:center;text-align:center;padding:5rem 2rem;}}
.hero-ornament{{color:#c88c3c;font-size:.9rem;letter-spacing:.3em;text-transform:uppercase;margin-bottom:1.5rem;}}
.hero h1{{font-family:'Playfair Display',serif;font-size:clamp(3rem,7vw,6.5rem);font-weight:400;line-height:1.05;margin-bottom:1.5rem;}}
.hero h1 em{{font-style:italic;color:#c88c3c;}}
.hero p{{font-size:1rem;color:rgba(240,232,216,.7);letter-spacing:.12em;text-transform:uppercase;margin-bottom:3rem;}}
.btn-amber{{display:inline-block;padding:.9rem 3rem;background:#c88c3c;color:#1a0f0a;font-family:'Lato',sans-serif;font-size:.85rem;font-weight:700;letter-spacing:.12em;text-transform:uppercase;border:none;cursor:pointer;transition:all .25s;}}
.btn-amber:hover{{background:#d8a04c;transform:translateY(-2px);}}
.btn-outline-amber{{display:inline-block;padding:.9rem 3rem;background:transparent;color:#c88c3c;border:1px solid #c88c3c;font-family:'Lato',sans-serif;font-size:.85rem;font-weight:700;letter-spacing:.12em;text-transform:uppercase;cursor:pointer;transition:all .25s;margin-left:1rem;}}
.btn-outline-amber:hover{{background:#c88c3c;color:#1a0f0a;}}
.menu{{padding:8rem 5rem;}}
.menu-header{{text-align:center;margin-bottom:5rem;}}
.menu-header h2{{font-family:'Playfair Display',serif;font-size:2.8rem;font-weight:400;margin-bottom:.5rem;}}
.menu-header p{{color:rgba(240,232,216,.5);letter-spacing:.1em;}}
.menu-grid{{display:grid;grid-template-columns:repeat(3,1fr);gap:2rem;max-width:1100px;margin:0 auto;}}
.menu-card{{background:#241510;border:1px solid rgba(200,140,60,.15);border-radius:4px;overflow:hidden;transition:all .3s;}}
.menu-card:hover{{border-color:#c88c3c;transform:translateY(-4px);}}
.menu-card img{{width:100%;aspect-ratio:4/3;object-fit:cover;filter:saturate(.85);}}
.menu-info{{padding:1.5rem;}}
.menu-info h3{{font-family:'Playfair Display',serif;font-size:1.15rem;margin-bottom:.3rem;}}
.menu-info p{{color:rgba(240,232,216,.55);font-size:.88rem;margin-bottom:.8rem;}}
.menu-price{{color:#c88c3c;font-weight:700;font-size:1rem;}}
.about{{background:#241510;padding:8rem 5rem;display:grid;grid-template-columns:1fr 1fr;gap:5rem;align-items:center;}}
.about img{{width:100%;border-radius:2px;box-shadow:0 20px 60px rgba(0,0,0,.5);}}
.about-text h2{{font-family:'Playfair Display',serif;font-size:2.5rem;font-weight:400;margin-bottom:1.5rem;line-height:1.2;}}
.about-text p{{color:rgba(240,232,216,.65);margin-bottom:1rem;}}
.reserve{{padding:8rem 5rem;text-align:center;}}
.reserve h2{{font-family:'Playfair Display',serif;font-size:2.5rem;font-weight:400;margin-bottom:1rem;}}
.reserve p{{color:rgba(240,232,216,.55);margin-bottom:3rem;}}
.reserve-form{{display:grid;grid-template-columns:repeat(3,1fr);gap:1rem;max-width:700px;margin:0 auto 1.5rem;}}
.reserve-form input, .reserve-form select{{background:transparent;border:1px solid rgba(200,140,60,.3);color:#f0e8d8;padding:.9rem 1rem;font-family:'Lato',sans-serif;font-size:.9rem;outline:none;transition:border-color .3s;border-radius:2px;}}
.reserve-form input:focus, .reserve-form select:focus{{border-color:#c88c3c;}}
footer{{padding:2.5rem 5rem;border-top:1px solid rgba(200,140,60,.15);display:flex;justify-content:space-between;font-size:.82rem;color:rgba(240,232,216,.35);align-items:center;}}
@media(max-width:900px){{header,.menu,.about,.reserve,footer{{padding-left:1.5rem;padding-right:1.5rem;}}.menu-grid,.about,.reserve-form{{grid-template-columns:1fr;}}}}
</style>
</head>
<body>
<header>
  <div class="logo">{title}</div>
  <nav><a href="#menu">Menu</a><a href="#about">Story</a><a href="#reserve" class="reserve-btn">Reserve</a></nav>
</header>
<div class="hero">
  <p class="hero-ornament">✦ Farm to Table · Est. 2016 ✦</p>
  <h1>Where Fire Meets<br><em>Flavour</em></h1>
  <p>Wood-fired cuisine · Locally sourced · Seasonally inspired</p>
  <button class="btn-amber">Reserve a Table</button>
  <button class="btn-outline-amber">View Menu</button>
</div>
<div class="menu" id="menu">
  <div class="menu-header">
    <h2>Tonight's Menu</h2>
    <p>Each dish crafted fresh with seasonal ingredients</p>
  </div>
  <div class="menu-grid">
    <div class="menu-card"><img src="https://picsum.photos/id/292/400/300" alt="Steak"><div class="menu-info"><h3>Oak-Fired Rib-Eye</h3><p>28-day dry-aged, bone marrow butter, charred leeks</p><div class="menu-price">$58</div></div></div>
    <div class="menu-card"><img src="https://picsum.photos/id/493/400/300" alt="Salmon"><div class="menu-info"><h3>Cedar Plank Salmon</h3><p>Wild Pacific, dill crème fraîche, pickled cucumber</p><div class="menu-price">$42</div></div></div>
    <div class="menu-card"><img src="https://picsum.photos/id/431/400/300" alt="Dessert"><div class="menu-info"><h3>Dark Chocolate Tart</h3><p>Valrhona 70%, salted caramel, hazelnut praline</p><div class="menu-price">$16</div></div></div>
  </div>
</div>
<div class="about" id="about">
  <img src="https://picsum.photos/id/1078/600/500" alt="Chef">
  <div class="about-text">
    <h2>Rooted in the Land,<br>Raised by Fire</h2>
    <p>Chef Marco Vidal opened {title} with a single belief: the best food comes from the best ingredients, simply prepared over open flame.</p>
    <p>We source exclusively from farms within 100 miles, changing our menu with every season to honour what the land provides.</p>
    <br><button class="btn-amber">Meet the Team</button>
  </div>
</div>
<div class="reserve" id="reserve">
  <h2>Reserve Your Table</h2>
  <p>We recommend booking in advance — our dining room fills quickly</p>
  <div class="reserve-form">
    <input type="text" placeholder="Full name">
    <input type="date">
    <select><option>2 guests</option><option>3 guests</option><option>4 guests</option><option>5+ guests</option></select>
  </div>
  <button class="btn-amber">Request Reservation</button>
</div>
<footer>
  <span>{title} &copy; {datetime.now().year}</span>
  <span>42 Hearth Street · hello@hearth.com</span>
</footer>
</body>
</html>"""

    # ── 5. Photography Studio ─────────────────────────────────────────────────
    @staticmethod
    def photography(title="Lens & Light Studio"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=Cormorant+Garamond:ital,wght@0,400;0,600;1,400&family=Montserrat:wght@300;400;500&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
body{{font-family:'Montserrat',sans-serif;background:#f9f7f5;color:#1a1510;line-height:1.75;}}
header{{padding:1.2rem 5rem;background:#f9f7f5;border-bottom:1px solid #e8e2da;position:sticky;top:0;z-index:99;display:grid;grid-template-columns:1fr auto 1fr;align-items:center;}}
nav.left{{display:flex;gap:2.5rem;}}
nav.right{{display:flex;gap:2.5rem;justify-content:flex-end;}}
nav a{{color:#8a7a6a;text-decoration:none;font-size:.8rem;letter-spacing:.1em;text-transform:uppercase;transition:color .3s;font-weight:500;}}
nav a:hover{{color:#1a1510;}}
.logo-center{{font-family:'Cormorant Garamond',serif;font-size:1.7rem;text-align:center;letter-spacing:.08em;color:#1a1510;}}
.hero{{display:grid;grid-template-columns:1fr 1fr;min-height:90vh;}}
.hero-left{{padding:5rem;display:flex;flex-direction:column;justify-content:center;border-right:1px solid #e8e2da;}}
.hero-tag{{font-size:.7rem;letter-spacing:.25em;text-transform:uppercase;color:#b8927a;margin-bottom:2rem;display:flex;align-items:center;gap:.8rem;}}
.hero-tag::before{{content:'';width:30px;height:1px;background:#b8927a;}}
.hero-left h1{{font-family:'Cormorant Garamond',serif;font-size:clamp(3rem,5vw,5.5rem);font-weight:400;line-height:1.05;margin-bottom:1.8rem;}}
.hero-left h1 em{{font-style:italic;color:#b8927a;}}
.hero-left p{{color:#8a7a6a;font-size:.95rem;max-width:360px;margin-bottom:2.5rem;}}
.hero-right{{background:url('https://picsum.photos/id/823/800/900') center/cover;}}
.btn-dark{{display:inline-block;padding:.9rem 2.5rem;background:#1a1510;color:#f9f7f5;font-family:'Montserrat',sans-serif;font-size:.78rem;font-weight:600;letter-spacing:.15em;text-transform:uppercase;border:none;cursor:pointer;transition:all .25s;}}
.btn-dark:hover{{background:#3a2a20;}}
.btn-line{{display:inline-block;padding:.9rem 2.5rem;background:transparent;color:#1a1510;border:1px solid #1a1510;font-family:'Montserrat',sans-serif;font-size:.78rem;font-weight:600;letter-spacing:.15em;text-transform:uppercase;cursor:pointer;transition:all .25s;margin-left:1rem;}}
.btn-line:hover{{background:#1a1510;color:#f9f7f5;}}
.gallery{{padding:5rem;}}
.gallery h2{{font-family:'Cormorant Garamond',serif;font-size:2.5rem;font-weight:400;text-align:center;margin-bottom:4rem;}}
.masonry{{columns:3;gap:1.2rem;}}
.masonry-item{{break-inside:avoid;margin-bottom:1.2rem;overflow:hidden;cursor:pointer;}}
.masonry-item img{{width:100%;display:block;transition:transform .5s,filter .5s;filter:saturate(.9);}}
.masonry-item:hover img{{transform:scale(1.03);filter:saturate(1.1);}}
.services{{background:#1a1510;color:#f9f7f5;padding:7rem 5rem;}}
.services h2{{font-family:'Cormorant Garamond',serif;font-size:2.5rem;font-weight:400;text-align:center;margin-bottom:4rem;}}
.services-grid{{display:grid;grid-template-columns:repeat(3,1fr);gap:3rem;max-width:1000px;margin:0 auto;text-align:center;}}
.srv h3{{font-family:'Cormorant Garamond',serif;font-size:1.4rem;font-weight:600;margin-bottom:.8rem;color:#b8927a;}}
.srv p{{color:rgba(249,247,245,.6);font-size:.88rem;}}
.srv .srv-price{{color:#b8927a;font-size:1.1rem;margin-top:.8rem;}}
.contact{{padding:7rem 5rem;display:grid;grid-template-columns:1fr 1fr;gap:5rem;align-items:start;}}
.contact-left h2{{font-family:'Cormorant Garamond',serif;font-size:2.5rem;font-weight:400;margin-bottom:1rem;line-height:1.2;}}
.contact-left h2 em{{font-style:italic;color:#b8927a;}}
.contact-left p{{color:#8a7a6a;}}
.c-form{{display:flex;flex-direction:column;gap:1.2rem;}}
.c-form input,.c-form textarea{{background:transparent;border:none;border-bottom:1px solid #e8e2da;color:#1a1510;padding:.7rem 0;font-family:'Montserrat',sans-serif;font-size:.9rem;outline:none;transition:border-color .3s;}}
.c-form input:focus,.c-form textarea:focus{{border-bottom-color:#b8927a;}}
footer{{padding:2rem 5rem;border-top:1px solid #e8e2da;display:flex;justify-content:space-between;font-size:.78rem;color:#8a7a6a;}}
@media(max-width:900px){{header{{grid-template-columns:auto;padding:1rem 1.5rem;}}.hero{{grid-template-columns:1fr;}}.hero-right{{height:300px;}}.gallery,.services,.contact{{padding:4rem 1.5rem;}}.masonry{{columns:2;}}.services-grid,.contact{{grid-template-columns:1fr;}}nav.left,nav.right{{display:none;}}}}
</style>
</head>
<body>
<header>
  <nav class="left"><a href="#gallery">Portfolio</a><a href="#services">Services</a></nav>
  <div class="logo-center">{title}</div>
  <nav class="right"><a href="#about">About</a><a href="#contact">Contact</a></nav>
</header>
<div class="hero">
  <div class="hero-left">
    <div class="hero-tag">Portrait & Editorial Photography</div>
    <h1>Capturing<br><em>Moments</em><br>That Last</h1>
    <p>Award-winning photography studio specialising in portrait, editorial, and commercial work.</p>
    <div>
      <button class="btn-dark">View Portfolio</button>
      <button class="btn-line">Book a Session</button>
    </div>
  </div>
  <div class="hero-right"></div>
</div>
<div class="gallery" id="gallery">
  <h2>Selected Work</h2>
  <div class="masonry">
    <div class="masonry-item"><img src="https://picsum.photos/id/1027/400/600" alt="Portrait"></div>
    <div class="masonry-item"><img src="https://picsum.photos/id/823/400/300" alt="Editorial"></div>
    <div class="masonry-item"><img src="https://picsum.photos/id/1003/400/500" alt="Lifestyle"></div>
    <div class="masonry-item"><img src="https://picsum.photos/id/1074/400/350" alt="Commercial"></div>
    <div class="masonry-item"><img src="https://picsum.photos/id/659/400/600" alt="Nature"></div>
    <div class="masonry-item"><img src="https://picsum.photos/id/399/400/300" alt="Fashion"></div>
  </div>
</div>
<div class="services" id="services">
  <h2>Sessions & Pricing</h2>
  <div class="services-grid">
    <div class="srv"><h3>Portrait</h3><p>2-hour studio session with 30 edited high-res images and full usage licence.</p><div class="srv-price">From $450</div></div>
    <div class="srv"><h3>Editorial</h3><p>Full-day shoot for magazines, brands, and creative campaigns.</p><div class="srv-price">From $1,200</div></div>
    <div class="srv"><h3>Commercial</h3><p>Product and advertising photography with licensing packages available.</p><div class="srv-price">Quote</div></div>
  </div>
</div>
<div class="contact" id="contact">
  <div class="contact-left">
    <h2>Let's Create<br>Something <em>Beautiful</em></h2>
    <p>Tell me about your project and let's find a date that works for you.</p>
  </div>
  <form class="c-form" onsubmit="return false;">
    <input type="text" placeholder="Your Name">
    <input type="email" placeholder="Email Address">
    <input type="text" placeholder="Session Type">
    <textarea rows="4" placeholder="Tell me about your vision..."></textarea>
    <button type="submit" class="btn-dark" style="width:fit-content;">Send Enquiry</button>
  </form>
</div>
<footer>
  <span>{title} &copy; {datetime.now().year}</span>
  <span>hello@lensandlight.com</span>
</footer>
</body>
</html>"""

    # ── 6. E-commerce / Store ─────────────────────────────────────────────────
    @staticmethod
    def ecommerce(title="Shop Modern"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=Nunito:wght@300;400;600;700;800&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
body{{font-family:'Nunito',sans-serif;background:#fff;color:#1a1a2e;line-height:1.6;}}
header{{padding:1rem 4rem;background:#fff;border-bottom:1px solid #f0f0f0;position:sticky;top:0;z-index:99;display:flex;justify-content:space-between;align-items:center;}}
.logo{{font-weight:800;font-size:1.4rem;color:#e94560;}}
nav a{{color:#4a4a6a;text-decoration:none;margin-left:2rem;font-size:.9rem;font-weight:600;transition:color .3s;}}
nav a:hover{{color:#e94560;}}
.cart-btn{{background:#e94560;color:#fff;padding:.55rem 1.3rem;border-radius:50px;font-weight:700;font-size:.85rem;}}
.hero{{background:linear-gradient(135deg,#fff5f7 0%,#ffeef2 100%);padding:6rem 4rem;display:grid;grid-template-columns:1fr 1fr;gap:4rem;align-items:center;}}
.hero-text .tag{{background:#ffd6de;color:#e94560;padding:.3rem 1rem;border-radius:50px;font-size:.8rem;font-weight:700;display:inline-block;margin-bottom:1.5rem;}}
.hero-text h1{{font-size:clamp(2.5rem,5vw,4rem);font-weight:800;line-height:1.1;margin-bottom:1.2rem;}}
.hero-text h1 span{{color:#e94560;}}
.hero-text p{{color:#6a6a8a;margin-bottom:2rem;font-size:1.05rem;}}
.hero-btns{{display:flex;gap:1rem;flex-wrap:wrap;}}
.btn-red{{background:#e94560;color:#fff;padding:.85rem 2rem;border:none;border-radius:50px;font-family:'Nunito',sans-serif;font-size:.95rem;font-weight:700;cursor:pointer;transition:all .25s;box-shadow:0 4px 20px rgba(233,69,96,.3);}}
.btn-red:hover{{transform:translateY(-2px);box-shadow:0 8px 30px rgba(233,69,96,.4);}}
.btn-border{{background:transparent;color:#1a1a2e;padding:.85rem 2rem;border:2px solid #1a1a2e;border-radius:50px;font-family:'Nunito',sans-serif;font-size:.95rem;font-weight:700;cursor:pointer;transition:all .25s;}}
.btn-border:hover{{background:#1a1a2e;color:#fff;}}
.hero-img{{border-radius:20px;overflow:hidden;box-shadow:0 20px 60px rgba(233,69,96,.15);}}
.hero-img img{{width:100%;display:block;}}
.featured{{padding:5rem 4rem;}}
.featured h2{{font-size:1.8rem;font-weight:800;margin-bottom:2.5rem;}}
.products{{display:grid;grid-template-columns:repeat(auto-fit,minmax(240px,1fr));gap:2rem;}}
.product{{border-radius:16px;overflow:hidden;border:1px solid #f0f0f0;transition:all .3s;cursor:pointer;}}
.product:hover{{box-shadow:0 12px 40px rgba(0,0,0,.1);transform:translateY(-4px);}}
.product-img{{aspect-ratio:1;overflow:hidden;}}
.product-img img{{width:100%;height:100%;object-fit:cover;transition:transform .4s;}}
.product:hover .product-img img{{transform:scale(1.06);}}
.product-info{{padding:1.2rem;}}
.product-info h3{{font-size:1rem;font-weight:700;margin-bottom:.3rem;}}
.product-info .price{{color:#e94560;font-weight:800;font-size:1.1rem;}}
.product-info .old-price{{text-decoration:line-through;color:#b0b0c0;font-size:.85rem;margin-left:.4rem;}}
.add-btn{{width:100%;margin-top:.8rem;background:#1a1a2e;color:#fff;border:none;border-radius:8px;padding:.7rem;font-family:'Nunito',sans-serif;font-weight:700;cursor:pointer;transition:background .25s;}}
.add-btn:hover{{background:#e94560;}}
.promo{{background:linear-gradient(135deg,#e94560,#ff6b8a);color:#fff;padding:5rem 4rem;text-align:center;}}
.promo h2{{font-size:2.5rem;font-weight:800;margin-bottom:1rem;}}
.promo p{{font-size:1.1rem;opacity:.9;margin-bottom:2rem;}}
.promo-code{{background:rgba(255,255,255,.2);border:2px dashed rgba(255,255,255,.5);display:inline-block;padding:.6rem 2rem;border-radius:8px;font-size:1.2rem;font-weight:800;letter-spacing:.1em;margin-bottom:2rem;}}
footer{{padding:2rem 4rem;background:#1a1a2e;color:rgba(255,255,255,.4);display:flex;justify-content:space-between;font-size:.85rem;}}
footer a{{color:inherit;text-decoration:none;}}
@media(max-width:768px){{header,.hero,.featured,.promo,footer{{padding-left:1.5rem;padding-right:1.5rem;}}.hero{{grid-template-columns:1fr;}}.hero-img{{display:none;}}}}
</style>
</head>
<body>
<header>
  <div class="logo">{title}</div>
  <nav><a href="#featured">Shop</a><a href="#about">About</a><a href="#contact">Contact</a><a href="#cart" class="cart-btn">🛒 Cart (0)</a></nav>
</header>
<div class="hero">
  <div class="hero-text">
    <span class="tag">🔥 New Collection — Summer 2024</span>
    <h1>Style That<br><span>Speaks</span><br>For Itself</h1>
    <p>Discover curated pieces that blend comfort, style, and sustainability. Made to be worn, made to last.</p>
    <div class="hero-btns">
      <button class="btn-red">Shop Now</button>
      <button class="btn-border">View Lookbook</button>
    </div>
  </div>
  <div class="hero-img"><img src="https://picsum.photos/id/996/600/600" alt="Fashion"></div>
</div>
<div class="featured" id="featured">
  <h2>Featured Products</h2>
  <div class="products">
    <div class="product"><div class="product-img"><img src="https://picsum.photos/id/118/400/400" alt="Product 1"></div><div class="product-info"><h3>Minimal Tee</h3><div><span class="price">$29</span><span class="old-price">$45</span></div><button class="add-btn">Add to Cart</button></div></div>
    <div class="product"><div class="product-img"><img src="https://picsum.photos/id/582/400/400" alt="Product 2"></div><div class="product-info"><h3>Classic Jacket</h3><div><span class="price">$89</span></div><button class="add-btn">Add to Cart</button></div></div>
    <div class="product"><div class="product-img"><img src="https://picsum.photos/id/696/400/400" alt="Product 3"></div><div class="product-info"><h3>Canvas Tote</h3><div><span class="price">$35</span><span class="old-price">$50</span></div><button class="add-btn">Add to Cart</button></div></div>
    <div class="product"><div class="product-img"><img src="https://picsum.photos/id/326/400/400" alt="Product 4"></div><div class="product-info"><h3>Summer Dress</h3><div><span class="price">$65</span></div><button class="add-btn">Add to Cart</button></div></div>
  </div>
</div>
<div class="promo">
  <h2>20% Off Your First Order</h2>
  <p>Use the code below at checkout and get 20% off everything.</p>
  <div class="promo-code">WELCOME20</div>
  <br><button class="btn-border" style="color:#fff;border-color:#fff;">Shop Now</button>
</div>
<footer>
  <span>&copy; {datetime.now().year} {title}. All rights reserved.</span>
  <div style="display:flex;gap:2rem;"><a href="#">Privacy</a><a href="#">Returns</a><a href="#">Contact</a></div>
</footer>
</body>
</html>"""

    # ── 8. Agency / Creative Studio ───────────────────────────────────────────
    @staticmethod
    def agency(title="Pixel & Craft Agency"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=Syne:wght@400;600;700;800&family=Epilogue:wght@300;400;500&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
body{{font-family:'Epilogue',sans-serif;background:#f0ece4;color:#161210;line-height:1.7;}}
header{{padding:1.2rem 4rem;background:#f0ece4;border-bottom:1px solid #ddd9d0;position:sticky;top:0;z-index:99;display:flex;justify-content:space-between;align-items:center;}}
.logo{{font-family:'Syne',sans-serif;font-weight:800;font-size:1.3rem;letter-spacing:-.02em;}}
nav a{{color:#6a6058;text-decoration:none;margin-left:2rem;font-size:.88rem;font-weight:500;transition:color .3s;}}
nav a:hover{{color:#161210;}}
.contact-cta{{background:#161210;color:#f0ece4!important;padding:.55rem 1.4rem;border-radius:50px;font-weight:600!important;}}
.hero{{padding:8rem 4rem 6rem;}}
.hero-label{{font-size:.75rem;letter-spacing:.2em;text-transform:uppercase;color:#9a8a7a;margin-bottom:2rem;}}
.hero h1{{font-family:'Syne',sans-serif;font-size:clamp(3rem,8vw,8rem);font-weight:800;line-height:.95;letter-spacing:-.04em;margin-bottom:3rem;}}
.hero h1 .line2{{display:block;margin-left:8rem;}}
.hero h1 .line2 em{{font-style:italic;font-weight:400;font-family:'Epilogue',serif;}}
.hero-footer{{display:flex;justify-content:space-between;align-items:flex-end;flex-wrap:wrap;gap:2rem;}}
.hero-footer p{{max-width:400px;color:#6a6058;}}
.btn-dark{{background:#161210;color:#f0ece4;padding:.9rem 2.5rem;border:none;border-radius:50px;font-family:'Epilogue',sans-serif;font-size:.9rem;font-weight:600;cursor:pointer;transition:all .25s;}}
.btn-dark:hover{{background:#3a2e28;transform:translateY(-2px);}}
.marquee-band{{background:#161210;color:#f0ece4;padding:1.5rem 0;overflow:hidden;border-top:1px solid #2a2420;border-bottom:1px solid #2a2420;}}
.marquee-track{{display:flex;gap:3rem;animation:scroll 20s linear infinite;white-space:nowrap;}}
@keyframes scroll{{from{{transform:translateX(0);}}to{{transform:translateX(-50%);}}}}
.marquee-item{{font-family:'Syne',sans-serif;font-size:1.1rem;font-weight:600;letter-spacing:.08em;text-transform:uppercase;display:flex;align-items:center;gap:3rem;}}
.marquee-item::after{{content:'✦';color:#9a8a7a;}}
.work{{padding:6rem 4rem;}}
.work-header{{display:flex;justify-content:space-between;align-items:flex-end;margin-bottom:4rem;}}
.work-header h2{{font-family:'Syne',sans-serif;font-size:2.5rem;font-weight:800;letter-spacing:-.03em;}}
.work-grid{{display:grid;grid-template-columns:2fr 1fr;gap:1.5rem;}}
.work-grid .row2{{display:grid;grid-template-columns:1fr;gap:1.5rem;}}
.work-item{{background:#e8e2d8;border-radius:12px;overflow:hidden;cursor:pointer;position:relative;}}
.work-item.tall{{aspect-ratio:3/4;}}
.work-item.short{{aspect-ratio:16/9;}}
.work-item img{{width:100%;height:100%;object-fit:cover;transition:transform .5s,filter .5s;filter:saturate(.85);}}
.work-item:hover img{{transform:scale(1.05);filter:saturate(1);}}
.work-caption{{position:absolute;bottom:0;left:0;right:0;padding:1.5rem;background:linear-gradient(to top,rgba(22,18,16,.8),transparent);color:#fff;opacity:0;transition:opacity .3s;}}
.work-item:hover .work-caption{{opacity:1;}}
.work-caption h3{{font-family:'Syne',sans-serif;font-size:1rem;font-weight:700;}}
.work-caption p{{font-size:.8rem;opacity:.7;}}
.about{{background:#161210;color:#f0ece4;padding:6rem 4rem;display:grid;grid-template-columns:1fr 1fr;gap:5rem;align-items:center;}}
.about h2{{font-family:'Syne',sans-serif;font-size:2.5rem;font-weight:800;letter-spacing:-.03em;line-height:1.1;margin-bottom:1.5rem;}}
.about h2 em{{font-family:'Epilogue',serif;font-style:italic;font-weight:300;color:#9a8a7a;}}
.about p{{color:rgba(240,236,228,.6);margin-bottom:1rem;}}
.stats-row{{display:grid;grid-template-columns:repeat(3,1fr);gap:2rem;margin-top:3rem;}}
.stat-val{{font-family:'Syne',sans-serif;font-size:2.5rem;font-weight:800;color:#c8a87a;}}
.stat-label{{font-size:.82rem;color:rgba(240,236,228,.4);}}
.contact-band{{padding:7rem 4rem;text-align:center;}}
.contact-band h2{{font-family:'Syne',sans-serif;font-size:clamp(2rem,6vw,5rem);font-weight:800;letter-spacing:-.04em;line-height:1;margin-bottom:2rem;}}
.contact-band h2 em{{font-family:'Epilogue',serif;font-style:italic;font-weight:300;}}
footer{{padding:2rem 4rem;border-top:1px solid #ddd9d0;display:flex;justify-content:space-between;font-size:.82rem;color:#9a8a7a;}}
@media(max-width:900px){{header,.hero,.work,.about,.contact-band,footer{{padding-left:1.5rem;padding-right:1.5rem;}}.work-grid,.about{{grid-template-columns:1fr;}}.hero h1 .line2{{margin-left:0;}}}}
</style>
</head>
<body>
<header>
  <div class="logo">{title}</div>
  <nav><a href="#work">Work</a><a href="#about">Studio</a><a href="#contact" class="contact-cta">Let's Talk</a></nav>
</header>
<div class="hero">
  <p class="hero-label">Creative Studio — Est. 2019</p>
  <h1>We Make<br><span class="line2">Brands <em>Unforgettable</em></span></h1>
  <div class="hero-footer">
    <p>Branding, digital design, and creative strategy for companies ready to stand out.</p>
    <button class="btn-dark">See Our Work</button>
  </div>
</div>
<div class="marquee-band">
  <div class="marquee-track">
    <span class="marquee-item">Brand Identity</span>
    <span class="marquee-item">Web Design</span>
    <span class="marquee-item">Motion</span>
    <span class="marquee-item">Strategy</span>
    <span class="marquee-item">Packaging</span>
    <span class="marquee-item">Brand Identity</span>
    <span class="marquee-item">Web Design</span>
    <span class="marquee-item">Motion</span>
    <span class="marquee-item">Strategy</span>
    <span class="marquee-item">Packaging</span>
  </div>
</div>
<div class="work" id="work">
  <div class="work-header">
    <h2>Selected Work</h2>
    <a href="#" style="color:#9a8a7a;font-size:.88rem;text-decoration:none;">View all →</a>
  </div>
  <div class="work-grid">
    <div class="work-item tall"><img src="https://picsum.photos/id/1051/600/800" alt="Project"><div class="work-caption"><h3>Aura Skincare</h3><p>Brand Identity · 2024</p></div></div>
    <div class="row2">
      <div class="work-item short"><img src="https://picsum.photos/id/175/600/338" alt="Project"><div class="work-caption"><h3>Velo App</h3><p>UI/UX · 2024</p></div></div>
      <div class="work-item short"><img src="https://picsum.photos/id/250/600/338" alt="Project"><div class="work-caption"><h3>Crux Coffee</h3><p>Packaging · 2023</p></div></div>
    </div>
  </div>
</div>
<div class="about" id="about">
  <div>
    <h2>Small Studio.<br><em>Big Thinking.</em></h2>
    <p>We're a tight-knit team of designers, strategists, and storytellers who believe good design changes minds and shapes culture.</p>
    <p>We've worked with startups, Fortune 500s, and everything in between — always with the same commitment to craft.</p>
    <div class="stats-row">
      <div><div class="stat-val">80+</div><div class="stat-label">Projects</div></div>
      <div><div class="stat-val">5yr</div><div class="stat-label">In Studio</div></div>
      <div><div class="stat-val">18</div><div class="stat-label">Awards</div></div>
    </div>
  </div>
  <img src="https://picsum.photos/id/3/600/500" alt="Studio" style="width:100%;border-radius:12px;">
</div>
<div class="contact-band" id="contact">
  <h2>Got a project?<br><em>Let's make it real.</em></h2>
  <button class="btn-dark" style="font-size:1rem;padding:1rem 3rem;">Start a Conversation →</button>
</div>
<footer>
  <span>{title} &copy; {datetime.now().year}</span>
  <div style="display:flex;gap:2rem;"><a href="#" style="color:inherit;text-decoration:none;">Privacy</a><a href="#" style="color:inherit;text-decoration:none;">Instagram</a></div>
</footer>
</body>
</html>"""

    # ── 9. Personal / CV / Resume ──────────────────────────────────────────────
    @staticmethod
    def personal_cv(title="Alex Johnson"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=IBM+Plex+Serif:ital,wght@0,400;0,600;1,400&family=IBM+Plex+Sans:wght@300;400;500;600&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
body{{font-family:'IBM Plex Sans',sans-serif;background:#0a0a12;color:#d8d8f0;line-height:1.75;}}
header{{padding:1.2rem 4rem;background:rgba(10,10,18,.9);backdrop-filter:blur(12px);position:sticky;top:0;z-index:99;border-bottom:1px solid rgba(120,120,200,.12);display:flex;justify-content:space-between;align-items:center;}}
.name-logo{{font-family:'IBM Plex Serif',serif;font-size:1.2rem;color:#a8a8e8;}}
nav a{{color:rgba(216,216,240,.5);text-decoration:none;margin-left:2rem;font-size:.88rem;transition:color .3s;}}
nav a:hover{{color:#a8a8e8;}}
.hero{{padding:8rem 4rem 6rem;display:grid;grid-template-columns:auto 1fr;gap:5rem;align-items:center;max-width:1000px;margin:0 auto;}}
.avatar{{width:180px;height:180px;border-radius:50%;object-fit:cover;border:3px solid rgba(120,120,200,.3);box-shadow:0 0 50px rgba(120,120,200,.2);flex-shrink:0;}}
.hero-text .role{{font-size:.8rem;letter-spacing:.2em;text-transform:uppercase;color:#7878c8;margin-bottom:1rem;}}
.hero-text h1{{font-family:'IBM Plex Serif',serif;font-size:clamp(2.5rem,4vw,4rem);font-weight:600;line-height:1.1;margin-bottom:1.2rem;}}
.hero-text p{{color:rgba(216,216,240,.6);max-width:480px;margin-bottom:2rem;}}
.hero-links{{display:flex;gap:1rem;flex-wrap:wrap;}}
.tag-link{{display:inline-block;padding:.45rem 1.1rem;border:1px solid rgba(120,120,200,.3);border-radius:50px;font-size:.82rem;color:#a8a8e8;text-decoration:none;transition:all .25s;}}
.tag-link:hover{{background:rgba(120,120,200,.15);border-color:#a8a8e8;}}
.section{{padding:5rem 4rem;max-width:1000px;margin:0 auto;}}
.section h2{{font-family:'IBM Plex Serif',serif;font-size:1.6rem;font-weight:600;margin-bottom:2.5rem;padding-bottom:1rem;border-bottom:1px solid rgba(120,120,200,.15);}}
.skills-grid{{display:grid;grid-template-columns:repeat(auto-fill,minmax(120px,1fr));gap:1rem;}}
.skill{{background:rgba(120,120,200,.08);border:1px solid rgba(120,120,200,.15);border-radius:8px;padding:.8rem;text-align:center;font-size:.88rem;color:rgba(216,216,240,.7);transition:all .25s;}}
.skill:hover{{background:rgba(120,120,200,.18);border-color:#a8a8e8;color:#d8d8f0;}}
.experience{{display:flex;flex-direction:column;gap:2.5rem;}}
.job{{border-left:2px solid rgba(120,120,200,.25);padding-left:2rem;position:relative;}}
.job::before{{content:'';position:absolute;left:-5px;top:8px;width:8px;height:8px;border-radius:50%;background:#7878c8;}}
.job-header{{display:flex;justify-content:space-between;align-items:flex-start;flex-wrap:wrap;gap:.5rem;margin-bottom:.5rem;}}
.job h3{{font-family:'IBM Plex Serif',serif;font-size:1.05rem;font-weight:600;}}
.job-date{{font-size:.82rem;color:#7878c8;}}
.job-company{{font-size:.9rem;color:rgba(216,216,240,.5);margin-bottom:.8rem;}}
.job p{{color:rgba(216,216,240,.6);font-size:.92rem;}}
.contact-section{{background:rgba(120,120,200,.05);border:1px solid rgba(120,120,200,.12);border-radius:16px;padding:3rem;text-align:center;max-width:600px;margin:0 auto 4rem;}}
.contact-section h2{{font-family:'IBM Plex Serif',serif;font-size:1.8rem;margin-bottom:1rem;}}
.contact-section p{{color:rgba(216,216,240,.55);margin-bottom:2rem;}}
.contact-section .email{{font-size:1.1rem;color:#a8a8e8;}}
footer{{padding:2rem 4rem;border-top:1px solid rgba(120,120,200,.12);text-align:center;color:rgba(216,216,240,.25);font-size:.82rem;}}
@media(max-width:700px){{header,.section{{padding-left:1.5rem;padding-right:1.5rem;}}.hero{{grid-template-columns:1fr;padding:4rem 1.5rem;}}.avatar{{width:120px;height:120px;}}}}
</style>
</head>
<body>
<header>
  <div class="name-logo">{title}</div>
  <nav><a href="#skills">Skills</a><a href="#experience">Experience</a><a href="#contact">Contact</a></nav>
</header>
<div class="hero">
  <img class="avatar" src="https://picsum.photos/id/1027/180/180" alt="{title}">
  <div class="hero-text">
    <div class="role">Senior Product Designer · UX Engineer</div>
    <h1>{title}</h1>
    <p>I design and build thoughtful digital products at the intersection of design and code. 6 years of experience shipping products loved by millions.</p>
    <div class="hero-links">
      <a href="#" class="tag-link">📄 Resume</a>
      <a href="#" class="tag-link">💼 LinkedIn</a>
      <a href="#" class="tag-link">🐙 GitHub</a>
      <a href="#" class="tag-link">🐦 Twitter</a>
    </div>
  </div>
</div>
<div class="section" id="skills">
  <h2>Skills & Tools</h2>
  <div class="skills-grid">
    <div class="skill">Figma</div><div class="skill">React</div><div class="skill">TypeScript</div>
    <div class="skill">Design Systems</div><div class="skill">UX Research</div><div class="skill">Framer</div>
    <div class="skill">Node.js</div><div class="skill">CSS / Tailwind</div><div class="skill">Prototyping</div>
    <div class="skill">Accessibility</div><div class="skill">SQL</div><div class="skill">Git</div>
  </div>
</div>
<div class="section" id="experience">
  <h2>Experience</h2>
  <div class="experience">
    <div class="job">
      <div class="job-header"><h3>Senior Product Designer</h3><span class="job-date">2022 — Present</span></div>
      <div class="job-company">Stripe · Remote</div>
      <p>Leading design for the Dashboard team. Shipped 3 major product surfaces, reduced support tickets by 40%, and built the component library used by 60+ designers.</p>
    </div>
    <div class="job">
      <div class="job-header"><h3>Product Designer</h3><span class="job-date">2020 — 2022</span></div>
      <div class="job-company">Linear · San Francisco</div>
      <p>Designed core product features from 0→1, including the roadmap view and integrations hub. Owned the mobile app redesign.</p>
    </div>
    <div class="job">
      <div class="job-header"><h3>UX Designer</h3><span class="job-date">2018 — 2020</span></div>
      <div class="job-company">Freelance · Various</div>
      <p>Partnered with 15+ startups on brand identity, web design, and product design. Built lasting client relationships and a diverse portfolio.</p>
    </div>
  </div>
</div>
<div class="section" id="contact">
  <div class="contact-section">
    <h2>Let's Work Together</h2>
    <p>I'm open to freelance projects, full-time roles, and interesting conversations.</p>
    <div class="email">hello@alexjohnson.design</div>
  </div>
</div>
<footer>&copy; {datetime.now().year} {title} · Made with care</footer>
</body>
</html>"""

    # ── 11. Fitness / Health ───────────────────────────────────────────────────
    @staticmethod
    def fitness(title="FitForge Studio"):
        return f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"><meta name="viewport" content="width=device-width,initial-scale=1.0">
<title>{title}</title>
<link href="https://fonts.googleapis.com/css2?family=Barlow+Condensed:wght@400;600;700;800;900&family=Barlow:wght@300;400;500&display=swap" rel="stylesheet">
<style>
*,*::before,*::after{{margin:0;padding:0;box-sizing:border-box;}}
:root{{--green:#00e676;--dark:#0c0c0c;}}
body{{font-family:'Barlow',sans-serif;background:var(--dark);color:#f0f0f0;line-height:1.7;}}
header{{padding:1.2rem 4rem;position:sticky;top:0;z-index:99;background:rgba(12,12,12,.92);backdrop-filter:blur(10px);border-bottom:1px solid rgba(0,230,118,.1);display:flex;justify-content:space-between;align-items:center;}}
.logo{{font-family:'Barlow Condensed',sans-serif;font-size:1.6rem;font-weight:900;letter-spacing:.06em;color:var(--green);text-transform:uppercase;}}
nav a{{color:rgba(240,240,240,.6);text-decoration:none;margin-left:2rem;font-size:.88rem;font-weight:500;text-transform:uppercase;letter-spacing:.08em;transition:color .3s;}}
nav a:hover{{color:var(--green);}}
.join-btn{{background:var(--green);color:#0c0c0c!important;padding:.55rem 1.4rem;font-weight:700!important;}}
.hero{{min-height:100vh;background:linear-gradient(rgba(12,12,12,.6),rgba(12,12,12,.8)),url('https://picsum.photos/id/416/1600/900') center/cover;display:flex;align-items:center;padding:6rem 4rem;}}
.hero-inner{{max-width:700px;}}
.hero-badge{{display:inline-block;background:rgba(0,230,118,.12);border:1px solid rgba(0,230,118,.3);color:var(--green);padding:.3rem 1rem;border-radius:4px;font-size:.75rem;font-weight:700;letter-spacing:.15em;text-transform:uppercase;margin-bottom:2rem;}}
.hero h1{{font-family:'Barlow Condensed',sans-serif;font-size:clamp(4rem,9vw,8rem);font-weight:900;line-height:.92;text-transform:uppercase;margin-bottom:1.5rem;}}
.hero h1 span{{color:var(--green);}}
.hero p{{font-size:1.1rem;color:rgba(240,240,240,.65);max-width:480px;margin-bottom:2.5rem;}}
.hero-stats{{display:flex;gap:3rem;margin-bottom:3rem;}}
.h-stat-num{{font-family:'Barlow Condensed',sans-serif;font-size:2.5rem;font-weight:900;color:var(--green);line-height:1;}}
.h-stat-lab{{font-size:.8rem;color:rgba(240,240,240,.4);text-transform:uppercase;letter-spacing:.1em;}}
.btn-green{{background:var(--green);color:#0c0c0c;padding:.9rem 2.5rem;border:none;border-radius:4px;font-family:'Barlow Condensed',sans-serif;font-size:1rem;font-weight:900;letter-spacing:.1em;text-transform:uppercase;cursor:pointer;transition:all .25s;}}
.btn-green:hover{{background:#33ff8f;transform:translateY(-2px);}}
.btn-outline-green{{background:transparent;color:var(--green);border:2px solid var(--green);padding:.9rem 2.5rem;border-radius:4px;font-family:'Barlow Condensed',sans-serif;font-size:1rem;font-weight:900;letter-spacing:.1em;text-transform:uppercase;cursor:pointer;transition:all .25s;margin-left:1rem;}}
.btn-outline-green:hover{{background:var(--green);color:#0c0c0c;}}
.programs{{padding:7rem 4rem;}}
.programs h2{{font-family:'Barlow Condensed',sans-serif;font-size:3rem;font-weight:900;text-transform:uppercase;margin-bottom:.5rem;}}
.programs .sub{{color:rgba(240,240,240,.45);margin-bottom:4rem;}}
.prog-grid{{display:grid;grid-template-columns:repeat(3,1fr);gap:1.5rem;}}
.prog{{background:#161616;border:1px solid #222;border-radius:8px;overflow:hidden;transition:all .3s;cursor:pointer;}}
.prog:hover{{border-color:var(--green);transform:translateY(-4px);}}
.prog-img{{aspect-ratio:16/9;overflow:hidden;}}
.prog-img img{{width:100%;height:100%;object-fit:cover;transition:transform .5s;filter:saturate(.8);}}
.prog:hover .prog-img img{{transform:scale(1.05);filter:saturate(1);}}
.prog-info{{padding:1.5rem;}}
.prog-tag{{font-size:.72rem;font-weight:700;letter-spacing:.15em;text-transform:uppercase;color:var(--green);margin-bottom:.5rem;}}
.prog-info h3{{font-family:'Barlow Condensed',sans-serif;font-size:1.3rem;font-weight:800;text-transform:uppercase;margin-bottom:.3rem;}}
.prog-info p{{color:rgba(240,240,240,.5);font-size:.88rem;}}
.trainers{{padding:7rem 4rem;background:#0f0f0f;}}
.trainers h2{{font-family:'Barlow Condensed',sans-serif;font-size:3rem;font-weight:900;text-transform:uppercase;text-align:center;margin-bottom:4rem;}}
.trainer-grid{{display:grid;grid-template-columns:repeat(auto-fit,minmax(200px,1fr));gap:2rem;max-width:900px;margin:0 auto;}}
.trainer img{{width:100%;aspect-ratio:3/4;object-fit:cover;margin-bottom:1rem;border-radius:4px;filter:grayscale(30%);transition:filter .3s;}}
.trainer:hover img{{filter:grayscale(0);}}
.trainer h3{{font-family:'Barlow Condensed',sans-serif;font-size:1.2rem;font-weight:800;text-transform:uppercase;margin-bottom:.2rem;}}
.trainer .spec{{color:var(--green);font-size:.82rem;font-weight:600;text-transform:uppercase;letter-spacing:.08em;}}
.cta-band{{background:var(--green);padding:5rem 4rem;text-align:center;}}
.cta-band h2{{font-family:'Barlow Condensed',sans-serif;font-size:3.5rem;font-weight:900;text-transform:uppercase;color:#0c0c0c;margin-bottom:1rem;}}
.cta-band p{{color:rgba(12,12,12,.65);margin-bottom:2rem;}}
.btn-dark{{background:#0c0c0c;color:var(--green);padding:1rem 3rem;border:none;border-radius:4px;font-family:'Barlow Condensed',sans-serif;font-size:1rem;font-weight:900;letter-spacing:.1em;text-transform:uppercase;cursor:pointer;transition:all .25s;}}
.btn-dark:hover{{background:#1a1a1a;}}
footer{{padding:2.5rem 4rem;border-top:1px solid #1a1a1a;display:flex;justify-content:space-between;font-size:.82rem;color:rgba(240,240,240,.25);}}
@media(max-width:900px){{header,.hero,.programs,.trainers,.cta-band,footer{{padding-left:1.5rem;padding-right:1.5rem;}}.prog-grid{{grid-template-columns:1fr;}}.hero-stats{{gap:1.5rem;}}}}
</style>
</head>
<body>
<header>
  <div class="logo">{title}</div>
  <nav><a href="#programs">Programs</a><a href="#trainers">Trainers</a><a href="#join" class="join-btn">Join Now</a></nav>
</header>
<div class="hero">
  <div class="hero-inner">
    <div class="hero-badge">⚡ Transform Your Body in 90 Days</div>
    <h1>Build The<br>Body You<br><span>Deserve.</span></h1>
    <p>Expert-designed programs, world-class coaches, and a community that pushes you further than you thought possible.</p>
    <div class="hero-stats">
      <div><div class="h-stat-num">12K+</div><div class="h-stat-lab">Members</div></div>
      <div><div class="h-stat-num">50+</div><div class="h-stat-lab">Programs</div></div>
      <div><div class="h-stat-num">98%</div><div class="h-stat-lab">Satisfaction</div></div>
    </div>
    <button class="btn-green">Start Free Trial</button>
    <button class="btn-outline-green">View Programs</button>
  </div>
</div>
<div class="programs" id="programs">
  <h2>Our Programs</h2>
  <p class="sub">From beginner to elite — find your perfect training program</p>
  <div class="prog-grid">
    <div class="prog"><div class="prog-img"><img src="https://picsum.photos/id/180/400/225" alt="Strength"></div><div class="prog-info"><div class="prog-tag">Strength</div><h3>Iron Foundation</h3><p>Progressive overload program for maximum muscle and strength gains.</p></div></div>
    <div class="prog"><div class="prog-img"><img src="https://picsum.photos/id/219/400/225" alt="HIIT"></div><div class="prog-info"><div class="prog-tag">Cardio</div><h3>Burn Protocol</h3><p>High-intensity interval training designed to shred fat and boost endurance.</p></div></div>
    <div class="prog"><div class="prog-img"><img src="https://picsum.photos/id/145/400/225" alt="Yoga"></div><div class="prog-info"><div class="prog-tag">Flexibility</div><h3>Mobility Flow</h3><p>Daily mobility and flexibility routines for recovery and injury prevention.</p></div></div>
  </div>
</div>
<div class="trainers" id="trainers">
  <h2>Meet Your Coaches</h2>
  <div class="trainer-grid">
    <div class="trainer"><img src="https://picsum.photos/id/1027/200/267" alt="Trainer"><h3>Marcus Webb</h3><div class="spec">Strength & Conditioning</div></div>
    <div class="trainer"><img src="https://picsum.photos/id/1005/200/267" alt="Trainer"><h3>Leila Nassar</h3><div class="spec">HIIT & Cardio</div></div>
    <div class="trainer"><img src="https://picsum.photos/id/1008/200/267" alt="Trainer"><h3>Tom Rivera</h3><div class="spec">Mobility & Yoga</div></div>
    <div class="trainer"><img src="https://picsum.photos/id/1012/200/267" alt="Trainer"><h3>Zoe Park</h3><div class="spec">Nutrition Coach</div></div>
  </div>
</div>
<div class="cta-band" id="join">
  <h2>Ready to Start?</h2>
  <p>Join 12,000+ members transforming their lives. First 7 days are completely free.</p>
  <button class="btn-dark">Start Free Trial Today</button>
</div>
<footer>
  <span>{title} &copy; {datetime.now().year}</span>
  <div style="display:flex;gap:2rem;"><a href="#" style="color:inherit;text-decoration:none;">Privacy</a><a href="#" style="color:inherit;text-decoration:none;">Terms</a></div>
</footer>
</body>
</html>"""


# ============================================================================
# Gradio callback functions
# ============================================================================
generation_history = []
_current_uploads: List[Dict] = []


def handle_uploads(file_list) -> Tuple[str, str]:
    global _current_uploads
    _current_uploads = ImageProvider.process_uploads(file_list)

    if not _current_uploads:
        _current_uploads = []
        return "", "No images uploaded yet."

    thumbs = ""
    for i, img in enumerate(_current_uploads):
        role  = "Hero background" if i == 0 else f"Portfolio card {i}"
        thumbs += f"""
        <div style="display:inline-block;margin:6px;text-align:center;vertical-align:top;">
            <img src="{img['data_url']}"
                 style="width:120px;height:90px;object-fit:cover;
                        border-radius:8px;border:2px solid {'#667eea' if i==0 else '#334155'};">
            <br>
            <span style="font-size:11px;color:#cbd5e1;display:block;max-width:120px;
                         overflow:hidden;text-overflow:ellipsis;white-space:nowrap;">{img['name']}</span>
            <span style="font-size:10px;color:{'#667eea' if i==0 else '#94a3b8'};">{role} • {img['size_kb']} KB</span>
        </div>"""

    gallery_html = f"""
    <div style="background:#1e293b;padding:12px;border-radius:10px;border:1px solid #334155;min-height:60px;">
        <p style="color:#94a3b8;font-size:12px;margin-bottom:8px;">
            {len(_current_uploads)} image(s) uploaded — first image = hero, rest = portfolio cards
        </p>
        {thumbs}
    </div>"""

    status = (f"**{len(_current_uploads)} image(s) ready.** "
              f"First image → hero background. Remaining {len(_current_uploads)-1} → portfolio cards.")
    return gallery_html, status


def generate_html_page(title, description, style="modern"):
    if not title.strip() or not description.strip():
        return "", "Please provide both a title and a description!", ""
    try:
        html_output, palette = HTMLGenerationEngine.generate_page(
            title, description, style, uploaded_images=_current_uploads or None)
        img_note = (f"\n- Images: {len(_current_uploads)} local file(s) embedded as base64"
                    if _current_uploads else "\n- Images: Picsum placeholders (no uploads provided)")
        color_info = f"""**Color Palette:**
- Primary:    {palette['primary']}
- Secondary:  {palette['secondary']}
- Accent:     {palette['accent']}
- Background: {palette['background']}{img_note}"""
        generation_history.append({"title": title, "html": html_output,
                                    "timestamp": datetime.now().isoformat()})
        return html_output, "✅ HTML page generated successfully!", color_info
    except Exception as e:
        return "", f"Error: {str(e)}", ""


def get_color_suggestions(style):
    palette = ColorPaletteGenerator.generate_palette(style)
    return f"""**Palette for "{style}" style:**
- Primary:    {palette['primary']}
- Secondary:  {palette['secondary']}
- Accent:     {palette['accent']}
- Background: {palette['background']}
- Text:       {palette['text_primary']}"""


def apply_template_generic(title, description, style, label):
    html, status, info = generate_html_page(title, description, style)
    return title, description, style, html, status, info, f"✅ {label} loaded and generated!"


# ── Premium template loaders ──────────────────────────────────────────────────
def load_restaurant():
    html = PremiumTemplates.restaurant("The Hearth Kitchen")
    return html, "✅ Restaurant template loaded!", "**Template: Restaurant** — Playfair Display · Dark amber · Menu cards"

def load_photography():
    html = PremiumTemplates.photography("Lens & Light Studio")
    return html, "✅ Photography template loaded!", "**Template: Photography Studio** — Cormorant Garamond · Warm tones · Masonry gallery"

def load_ecommerce():
    html = PremiumTemplates.ecommerce("Shop Modern")
    return html, "✅ E-commerce template loaded!", "**Template: E-commerce Store** — Nunito · Coral red · Product grid with cart"

def load_agency():
    html = PremiumTemplates.agency("Pixel & Craft Agency")
    return html, "✅ Agency template loaded!", "**Template: Creative Agency** — Syne display · Scrolling marquee · Bold typography"

def load_personal_cv():
    html = PremiumTemplates.personal_cv("Alex Johnson")
    return html, "✅ Personal CV template loaded!", "**Template: Personal CV** — IBM Plex · Dark purple · Skills & experience timeline"

def load_fitness():
    html = PremiumTemplates.fitness("FitForge Studio")
    return html, "✅ Fitness template loaded!", "**Template: Fitness Studio** — Barlow Condensed · Neon green · Programs & trainers"


# ============================================================================
# UX Reviewer (unchanged from original)
# ============================================================================
class UXReviewer:
    WEIGHTS = {"Section Coverage": 2.5, "Accessibility": 2.5,
               "Design Balance": 2.5, "Responsiveness": 2.5}

    @staticmethod
    def _has(html, *patterns):
        h = html.lower()
        return all(p.lower() in h for p in patterns)

    @staticmethod
    def _count(html, pattern):
        return len(re.findall(pattern, html, re.IGNORECASE))

    @staticmethod
    def _check_sections(html):
        good, warn, bad = [], [], []
        score = 10.0
        checks = [
            ("hero", '<section class="hero"', "Hero section", 2.0),
            ("features", 'class="features"', "Features section", 1.5),
            ("about", 'class="about"', "About section", 1.5),
            ("portfolio", 'class="portfolio"', "Portfolio section", 1.5),
            ("contact", 'class="contact"', "Contact section", 1.5),
            ("footer", "<footer", "Footer", 1.0),
            ("cta button", 'class="btn btn-primary"', "CTA button", 1.0),
        ]
        for key, pattern, label, deduct in checks:
            if UXReviewer._has(html, pattern):
                good.append(f"{label} is present")
            else:
                score -= deduct
                (bad if deduct >= 1.5 else warn).append(f"{label} is missing")
        return max(score, 0), good, warn, bad

    @staticmethod
    def _check_accessibility(html):
        good, warn, bad = [], [], []
        score = 10.0
        img_count = UXReviewer._count(html, r"<img\b")
        alt_count  = UXReviewer._count(html, r'<img\b[^>]*\balt\s*=\s*["\'][^"\']+["\']')
        empty_alt  = UXReviewer._count(html, r'alt\s*=\s*["\']["\']')
        if img_count == 0:
            warn.append("No images found"); score -= 1.0
        elif alt_count >= img_count - empty_alt:
            good.append(f"All {img_count} image(s) have alt text")
        else:
            bad.append(f"{img_count - alt_count} image(s) missing alt"); score -= min((img_count-alt_count)*1.0, 3.0)
        if UXReviewer._has(html, "<h1") and UXReviewer._has(html, "<h2"):
            good.append("Proper heading hierarchy used")
        elif UXReviewer._has(html, "<h1"):
            warn.append("Only h1 headings found"); score -= 1.0
        else:
            bad.append("No h1 heading found"); score -= 2.0
        if UXReviewer._has(html, 'lang="en"'):
            good.append('lang="en" is set correctly')
        else:
            warn.append('Missing lang attribute'); score -= 1.0
        if UXReviewer._has(html, '<meta name="description"'):
            good.append("Meta description present")
        else:
            warn.append("Meta description missing"); score -= 0.5
        return max(score, 0), good, warn, bad

    @staticmethod
    def _check_design(html):
        good, warn, bad = [], [], []
        score = 10.0
        anim = UXReviewer._count(html, r"animation\s*:")
        if anim == 0: warn.append("No CSS animations"); score -= 1.0
        elif anim <= 4: good.append(f"{anim} animation(s) — well balanced")
        else: warn.append(f"{anim} animations — may hurt perf"); score -= 1.5
        if UXReviewer._count(html, r"transition\s*:") >= 3:
            good.append("Smooth hover transitions applied")
        else:
            warn.append("Few hover transitions"); score -= 0.5
        if UXReviewer._has(html, ":root", "--primary"):
            good.append("CSS variables used for theming")
        else:
            warn.append("No CSS variables found"); score -= 1.0
        if UXReviewer._has(html, ".btn-primary") or UXReviewer._has(html, "btn-primary"):
            good.append("Primary CTA button style defined")
        else:
            bad.append("No primary button style found"); score -= 2.0
        if UXReviewer._has(html, ":hover"):
            good.append("Hover effects present")
        else:
            warn.append("No :hover effects"); score -= 1.0
        if UXReviewer._has(html, "box-shadow"):
            good.append("Box-shadows add visual depth")
        else:
            warn.append("No box-shadows found"); score -= 0.5
        return max(score, 0), good, warn, bad

    @staticmethod
    def _check_responsiveness(html):
        good, warn, bad = [], [], []
        score = 10.0
        if UXReviewer._has(html, 'name="viewport"', "width=device-width"):
            good.append("Viewport meta tag set correctly")
        else:
            bad.append("Viewport meta tag missing"); score -= 3.0
        mq = UXReviewer._count(html, r"@media\s*\(")
        if mq >= 2: good.append(f"{mq} media query breakpoints found")
        elif mq == 1: warn.append("Only 1 media query found"); score -= 1.0
        else: bad.append("No media queries found"); score -= 2.5
        if UXReviewer._has(html, "max-width"):
            good.append("max-width used on containers")
        else:
            warn.append("No max-width found"); score -= 1.0
        uses_flex = "display:flex" in html or "display: flex" in html
        uses_grid = "display:grid" in html or "display: grid" in html
        if uses_flex and uses_grid: good.append("Both Grid and Flexbox used")
        elif uses_flex: good.append("Flexbox layout used")
        elif uses_grid: good.append("CSS Grid layout used")
        else: bad.append("No Flexbox or Grid found"); score -= 2.0
        return max(score, 0), good, warn, bad

    @staticmethod
    def review(html):
        if not html or not html.strip():
            return "<p style='color:#f87171;padding:1rem;'>No HTML to review.</p>"
        sec_s, sec_g, sec_w, sec_b = UXReviewer._check_sections(html)
        acc_s, acc_g, acc_w, acc_b = UXReviewer._check_accessibility(html)
        des_s, des_g, des_w, des_b = UXReviewer._check_design(html)
        res_s, res_g, res_w, res_b = UXReviewer._check_responsiveness(html)
        overall = round(sum((s/10)*UXReviewer.WEIGHTS[c] for c, s in
            zip(["Section Coverage","Accessibility","Design Balance","Responsiveness"],
                [sec_s, acc_s, des_s, res_s])), 1)
        col = "#4ade80" if overall >= 8 else "#facc15" if overall >= 6 else "#fb923c" if overall >= 4 else "#f87171"
        lbl = "Excellent" if overall >= 8 else "Good" if overall >= 6 else "Needs Work" if overall >= 4 else "Poor"

        def item_html(icon, color, text):
            return f"<li style='margin:5px 0;color:{color};'><span style='margin-right:6px;'>{icon}</span>{text}</li>"

        def cat_block(title, s, g, w, b, icon):
            pct = round(s*10)
            bc  = "#4ade80" if s >= 7 else "#facc15" if s >= 5 else "#f87171"
            items = "".join([item_html("✔","#4ade80",x) for x in g] +
                            [item_html("⚠","#facc15",x) for x in w] +
                            [item_html("✖","#f87171",x) for x in b])
            return f"""<div style="background:#1e293b;border:1px solid #334155;border-radius:12px;padding:16px;margin-bottom:14px;">
                <div style="display:flex;justify-content:space-between;align-items:center;margin-bottom:8px;">
                    <span style="font-weight:700;color:#f1f5f9;font-size:15px;">{icon} {title}</span>
                    <span style="font-weight:700;color:{bc};font-size:15px;">{round(s,1)} / 10</span>
                </div>
                <div style="background:#0f172a;border-radius:6px;height:8px;margin-bottom:12px;">
                    <div style="background:{bc};width:{pct}%;height:8px;border-radius:6px;"></div>
                </div>
                <ul style="list-style:none;padding:0;margin:0;font-size:13px;">{items}</ul>
            </div>"""

        all_issues = sec_w+sec_b+acc_w+acc_b+des_w+des_b+res_w+res_b
        sugg_html  = "".join(f"<li style='margin:6px 0;color:#cbd5e1;font-size:13px;'><span style='color:#818cf8;font-weight:700;margin-right:6px;'>{i+1}.</span>{s}</li>"
                             for i, s in enumerate(all_issues[:5]))
        return f"""<div style="font-family:'Segoe UI',sans-serif;background:#0f172a;border-radius:16px;padding:20px;color:#f1f5f9;">
            <div style="text-align:center;margin-bottom:24px;">
                <div style="font-size:13px;color:#94a3b8;letter-spacing:1px;text-transform:uppercase;margin-bottom:4px;">AI UX Reviewer</div>
                <div style="font-size:52px;font-weight:900;color:{col};line-height:1;">{overall}</div>
                <div style="font-size:14px;color:#94a3b8;margin-top:2px;">/ 10 &nbsp;—&nbsp; <span style="color:{col};font-weight:600;">{lbl}</span></div>
            </div>
            {cat_block("Section Coverage", sec_s, sec_g, sec_w, sec_b, "🗂️")}
            {cat_block("Accessibility",    acc_s, acc_g, acc_w, acc_b, "♿")}
            {cat_block("Design Balance",   des_s, des_g, des_w, des_b, "🎨")}
            {cat_block("Responsiveness",   res_s, res_g, res_w, res_b, "📱")}
            <div style="background:#1e293b;border:1px solid #334155;border-radius:12px;padding:16px;">
                <div style="font-weight:700;color:#f1f5f9;font-size:15px;margin-bottom:10px;">💡 Top Suggestions</div>
                <ul style="list-style:none;padding:0;margin:0;">{sugg_html or "<li style='color:#4ade80;font-size:13px;'>No major issues — great work!</li>"}</ul>
            </div>
        </div>"""


# ============================================================================
# Gradio UI  —  Light theme with Inter/Nunito font
# ============================================================================
print("Loading HTML Quick-Styler v3 — Premium Templates Edition...")

_ocean_theme = gr.themes.Soft(
    primary_hue="blue",
    secondary_hue="cyan",
    neutral_hue="gray",
    font=[gr.themes.GoogleFont("Plus Jakarta Sans"), "ui-sans-serif", "sans-serif"],
    font_mono=[gr.themes.GoogleFont("JetBrains Mono"), "ui-monospace", "monospace"],
).set(
    body_background_fill="#ffffff",
    body_background_fill_dark="#ffffff",
    block_background_fill="#ffffff",
    block_border_color="#e5e7eb",
    block_shadow="0 1px 8px rgba(0,0,0,0.06)",
    block_radius="14px",
    button_primary_background_fill="linear-gradient(135deg,#1a6fc4,#0ea5e9)",
    button_primary_background_fill_hover="linear-gradient(135deg,#1558a0,#0284c7)",
    button_primary_text_color="#ffffff",
    button_secondary_background_fill="#ffffff",
    button_secondary_border_color="#d1d5db",
    button_secondary_text_color="#374151",
    input_background_fill="#ffffff",
    input_border_color="#d1d5db",
    input_border_color_focus="#1a6fc4",
    checkbox_background_color="#ffffff",
    slider_color="#1a6fc4",
)

with gr.Blocks(
    title="HTML Quick-Styler v3 — Premium Templates",
    theme=_ocean_theme,
    css="""
    @import url('https://fonts.googleapis.com/css2?family=Plus+Jakarta+Sans:wght@300;400;500;600;700;800&display=swap');

    /* ── Base: pure white everywhere ── */
    body, .gradio-container, .main, .wrap { font-family: 'Plus Jakarta Sans', sans-serif !important; background: #ffffff !important; }

    /* ── App-level header banner — blue ── */
    .gradio-container > .main > .wrap > .md:first-child,
    .gradio-container .app-header,
    .app > .header { background: linear-gradient(135deg,#1a6fc4,#0ea5e9) !important; border-radius: 14px !important; padding: 1.4rem 2rem !important; margin-bottom: 0.5rem !important; }

    /* ── Header text: white on blue banner ── */
    .gradio-container > .main > .wrap > .md:first-child h1,
    .gradio-container > .main > .wrap > .md:first-child h3,
    .gradio-container > .main > .wrap > .md:first-child p { color: #ffffff !important; }

    /* ── All other headings: dark for readability on white ── */
    h2, h3, h4 { color: #111827 !important; font-weight: 700 !important; }

    /* ── Tab bar: white bg, blue active indicator ── */
    .tab-nav { background: #ffffff !important; border-bottom: 2px solid #e5e7eb !important; }
    .tab-nav button { font-family: 'Plus Jakarta Sans', sans-serif !important; font-weight: 600 !important; color: #6b7280 !important; background: #ffffff !important; border-radius: 0 !important; }
    .tab-nav button.selected { color: #1a6fc4 !important; border-bottom: 3px solid #1a6fc4 !important; background: #ffffff !important; }
    .tab-nav button:hover { color: #1a6fc4 !important; background: #f0f7ff !important; }

    /* ── Blocks / panels: white with light grey border ── */
    .block, .form, .panel { background: #ffffff !important; border: 1px solid #e5e7eb !important; border-radius: 14px !important; box-shadow: 0 1px 8px rgba(0,0,0,0.05) !important; }

    /* ── Primary buttons: blue gradient ── */
    button.primary { border-radius: 10px !important; font-weight: 700 !important; box-shadow: 0 4px 14px rgba(14,105,190,0.25) !important; transition: transform 0.15s, box-shadow 0.15s !important; }
    button.primary:hover { transform: translateY(-2px) !important; box-shadow: 0 8px 22px rgba(14,105,190,0.35) !important; }

    /* ── Secondary buttons: white with grey border ── */
    button.secondary { background: #ffffff !important; border: 1.5px solid #d1d5db !important; color: #374151 !important; border-radius: 10px !important; font-weight: 600 !important; }
    button.secondary:hover { border-color: #1a6fc4 !important; color: #1a6fc4 !important; }

    /* ── Inputs: white bg, light border, blue focus ── */
    input, textarea, select { border-radius: 10px !important; border: 1.5px solid #d1d5db !important; background: #ffffff !important; font-family: 'Plus Jakarta Sans', sans-serif !important; color: #111827 !important; }
    input:focus, textarea:focus { border-color: #1a6fc4 !important; box-shadow: 0 0 0 3px rgba(26,111,196,0.12) !important; outline: none !important; }

    /* ── Labels: dark grey on white ── */
    label, .gr-label { font-weight: 600 !important; color: #374151 !important; font-size: 0.85rem !important; }

    /* ── Markdown prose: dark on white ── */
    .prose, .markdown-body, .gr-markdown { color: #1f2937 !important; }

    /* ── Dividers & scrollbars ── */
    hr { border-color: #e5e7eb !important; }
    ::-webkit-scrollbar { width: 6px; } ::-webkit-scrollbar-thumb { background: #d1d5db; border-radius: 3px; } ::-webkit-scrollbar-track { background: #f9fafb; }
    """
) as demo:

     gr.Markdown("""
# 🎨HTML Quick-Styler v3
### Professional Web Builder: AI-Powered Templates & UX Design
""")
     with gr.Tabs():

        # ── TAB 1: Generate Page ─────────────────────────────────────────────
        with gr.TabItem("🎯 Generate Page"):
            with gr.Row():
                with gr.Column(scale=1):
                    gr.Markdown("### Page Details")
                    page_title = gr.Textbox(label="Page Title", value="Modern Portfolio Website")
                    page_description = gr.Textbox(
                        label="Page Description & Sections",
                        value="Hero section with welcome message, features list with 3 items, about section, portfolio gallery, testimonials, contact form, and footer",
                        lines=4)
                    color_style = gr.Dropdown(
                        choices=["modern","vibrant","ocean","forest","sunset","luxury","creative","minimal"],
                        value="modern", label="Color Style")
                    gr.Markdown("""
### Upload Images *(optional)*
- **1st image** → hero background
- **2nd–7th images** → portfolio cards
""")
                    image_upload = gr.File(label="Choose Images", file_count="multiple",
                                           file_types=["image"], type="filepath")
                    upload_gallery = gr.HTML(value="<p style='color:#94a3b8;font-size:13px;padding:8px;'>No images uploaded yet.</p>")
                    upload_status  = gr.Markdown("No images uploaded yet.")
                    image_upload.change(fn=handle_uploads, inputs=[image_upload],
                                        outputs=[upload_gallery, upload_status])
                    generate_btn = gr.Button("✨ Generate HTML", size="lg", variant="primary")

                with gr.Column(scale=1):
                    gr.Markdown("### Output")
                    html_output = gr.Code(language="html", label="Generated HTML/CSS", lines=20)
                    status_msg  = gr.Markdown()
                    color_info  = gr.Markdown()

            generate_btn.click(fn=generate_html_page,
                                inputs=[page_title, page_description, color_style],
                                outputs=[html_output, status_msg, color_info])

        # ── TAB 2: Live Preview ───────────────────────────────────────────────
        with gr.TabItem("👁️ Live Preview"):
            gr.Markdown("### Preview — click **Load Preview** after generating or loading a template")
            preview_btn    = gr.Button("🔄 Load Preview", size="lg", variant="primary")
            preview_output = gr.HTML(value="<p style='text-align:center;color:#9ca3af;padding:60px;font-family:Plus Jakarta Sans,sans-serif;'>No preview loaded yet. Generate or load a template first.</p>")

            def show_preview(html_text):
                if not html_text or not html_text.strip():
                    return "<p style='text-align:center;color:#9ca3af;padding:60px;font-family:Plus Jakarta Sans,sans-serif;'>No HTML provided</p>"
                return html_text

            preview_btn.click(fn=show_preview, inputs=[html_output], outputs=[preview_output])

        # ── TAB 3: Color Palettes ─────────────────────────────────────────────
        with gr.TabItem("🎨 Color Palettes"):
            gr.Markdown("### Explore Color Schemes")
            with gr.Row():
                style_selector = gr.Radio(
                    choices=["modern","vibrant","ocean","forest","sunset","luxury","creative","minimal"],
                    value="modern", label="Choose Style")
                palette_btn = gr.Button("Get Palette", size="lg", variant="primary")
            palette_display = gr.Markdown()
            palette_btn.click(fn=get_color_suggestions, inputs=[style_selector], outputs=[palette_display])

            # ── TAB 4: AI UX Reviewer ─────────────────────────────────────────────
        with gr.TabItem("🧠 AI UX Reviewer"):
            gr.Markdown("""### AI UX Reviewer
Generate a page first, then click **Run UX Review** to score it across 4 UX dimensions.
""")
            ux_btn    = gr.Button("🔍 Run UX Review", size="lg", variant="primary")
            ux_report = gr.HTML(value="<p style='color:#9ca3af;padding:1rem;font-family:Plus Jakarta Sans,sans-serif;'>No review yet. Generate a page first.</p>")
            ux_btn.click(fn=UXReviewer.review, inputs=[html_output], outputs=[ux_report])

        # ── TAB 5: Templates ──────────────────────────────────────────────────
        with gr.TabItem("📋 Templates"):
            gr.Markdown("""### Templates — Click to Load Instantly
Each button loads a complete HTML page into the code editor. Then go to **👁️ Live Preview → Load Preview** to see it rendered.
""")

            # ── Classic Templates (row 1) ────────────────────────────────────
            gr.Markdown("#### Classic Templates")
            with gr.Row():
                with gr.Column():
                    gr.Markdown("**🗂️ Portfolio Website**\nHero, Gallery, About, Contact, Footer")
                    portfolio_orig_btn = gr.Button("Use Portfolio Template", variant="primary")
                with gr.Column():
                    gr.Markdown("**📰 Blog / Magazine**\nEditorial hero, article grid, newsletter strip")
                    blog_orig_btn = gr.Button("Use Blog / Magazine Template", variant="primary")
                with gr.Column():
                    gr.Markdown("**🎤 Event / Conference**\nCountdown, speakers, schedule, tickets")
                    event_orig_btn = gr.Button("Use Event Template", variant="primary")

            gr.Markdown("---")

            # ── Premium Templates (6 only) ───────────────────────────────────
            gr.Markdown("#### ✨ Premium Templates")
            with gr.Row():
                with gr.Column():
                    gr.Markdown("**📷 Photography Studio**\nCormorant Garamond · Warm editorial tones · Masonry gallery")
                    photography_btn = gr.Button("Load Photography Studio", variant="primary")
                with gr.Column():
                    gr.Markdown("**🍽️ Restaurant**\nPlayfair Display · Dark amber · Menu cards & reservation form")
                    restaurant_btn = gr.Button("Load Restaurant", variant="primary")
                with gr.Column():
                    gr.Markdown("**🛍️ E-commerce Store**\nNunito · Coral red · Product grid with add-to-cart")
                    ecommerce_btn = gr.Button("Load E-commerce", variant="primary")

            with gr.Row():
                with gr.Column():
                    gr.Markdown("**🎨 Creative Agency**\nSyne display · Scrolling marquee · Bold editorial layout")
                    agency_btn = gr.Button("Load Creative Agency", variant="primary")
                with gr.Column():
                    gr.Markdown("**👤 Personal CV**\nIBM Plex · Soft purple · Skills grid & experience timeline")
                    cv_btn = gr.Button("Load Personal CV", variant="primary")
                with gr.Column():
                    gr.Markdown("**💪 Fitness Studio**\nBarlow Condensed · Neon green · Programs & trainer cards")
                    fitness_btn = gr.Button("Load Fitness Studio", variant="primary")

            template_status = gr.Markdown()

            # Shared output lists
            _tmpl_base = [page_title, page_description, color_style,
                          html_output, status_msg, color_info, template_status]
            _tmpl_out  = [html_output, status_msg, color_info]

            # Classic template callbacks
            def apply_portfolio_orig():
                return apply_template_generic(
                    "Portfolio Website",
                    "Hero section with welcome banner, portfolio gallery with 6 items, about section, testimonials, contact form, footer",
                    "modern", "Portfolio template")
            def apply_blog_orig():
                return apply_template_generic(
                    "The Daily Brief",
                    "Hero section, features, about section, testimonials, contact form, footer",
                    "minimal", "Blog / Magazine template")
            def apply_event_orig():
                return apply_template_generic(
                    "TechConf 2025",
                    "Hero section with banner, features showcase, pricing plans, testimonials, contact form, footer",
                    "vibrant", "Event template")

            portfolio_orig_btn.click(fn=apply_portfolio_orig, inputs=[], outputs=_tmpl_base)
            blog_orig_btn.click(     fn=apply_blog_orig,      inputs=[], outputs=_tmpl_base)
            event_orig_btn.click(    fn=apply_event_orig,     inputs=[], outputs=_tmpl_base)

            # Premium template callbacks (6 only)
            photography_btn.click(fn=load_photography,   inputs=[], outputs=_tmpl_out)
            restaurant_btn.click( fn=load_restaurant,    inputs=[], outputs=_tmpl_out)
            ecommerce_btn.click(  fn=load_ecommerce,     inputs=[], outputs=_tmpl_out)
            agency_btn.click(     fn=load_agency,        inputs=[], outputs=_tmpl_out)
            cv_btn.click(         fn=load_personal_cv,   inputs=[], outputs=_tmpl_out)
            fitness_btn.click(    fn=load_fitness,       inputs=[], outputs=_tmpl_out)



        # ── TAB 6: Guide ──────────────────────────────────────────────────────
        with gr.TabItem("📚 Guide"):
            gr.Markdown("""
# 📚 HTML Quick-Styler — Complete User Guide

---

## 🚀 Getting Started

**HTML Quick-Styler v3** lets you create beautiful, production-ready HTML pages in seconds — no coding required.
You can either **generate** a page from scratch using AI, or **load a hand-crafted template** and preview it instantly.

---

## 🗂️ Tab-by-Tab Walkthrough

### 🎯 Tab 1 — Generate Page

This is your main workspace for creating new HTML pages using AI.

**Step-by-step:**
1. **Page Title** — Enter a short, descriptive name (e.g. `"My Photography Portfolio"` or `"TechConf 2025"`).
2. **Page Description & Sections** — Describe what sections you want. Be specific! The more detail you give, the better the output.
   - ✅ Good: `"Hero with tagline, 3-column features grid, about section with team photo, testimonials from 3 clients, contact form, footer"`
   - ❌ Too vague: `"Make a website"`
3. **Color Style** — Choose from 8 built-in palettes (see Color Palettes tab for previews).
4. **Upload Images** *(optional)* — Upload your own photos to embed them into the page as base64.
5. Click **✨ Generate HTML** — the full HTML + CSS code appears in the Output panel on the right.

## 🖼️ Image Upload Guide

You can embed your own photos directly into generated pages (stored as base64 — no external server needed).

| Upload Order | Where It Appears |
|---|---|
| 1st image | Hero section background |
| 2nd image | Portfolio card 1 |
| 3rd image | Portfolio card 2 |
| 4th–7th images | Portfolio cards 3–6 |

**Supported formats:** JPG, PNG, WebP, GIF
**Tip:** Keep hero images under 500 KB for faster loading. The tool shows you each file's size after upload.

---

> 💡 **Tip:** After generating, go to the **👁️ Live Preview** tab and click **Load Preview** to see the page rendered in your browser.

---

### 👁️ Tab 2 — Live Preview

Renders your current HTML code as a live webpage directly inside the app.

- Works with both **generated pages** and **loaded templates**
- Click **🔄 Load Preview** any time you want to refresh the view
- The preview is fully interactive — links, hover effects, and animations all work
- Use your browser's **Right Click → View Page Source** on the preview to inspect the HTML

> 💡 **Tip:** If you edited the code manually in the Output panel, click Load Preview again to see your changes.

---

### 🎨 Tab 3 — Color Palettes

Explore and compare the 8 built-in color schemes before generating a page.

| Palette | Mood | Best For |
|---|---|---|
| **modern** | Cool blue-purple | Portfolios, tech startups |
| **vibrant** | Warm coral-red | Creative agencies, lifestyle |
| **ocean** | Deep sea blue | Finance, corporate, SaaS |
| **forest** | Natural green | Eco brands, health, wellness |
| **sunset** | Warm orange | Food, travel, photography |
| **luxury** | Black & gold | Premium brands, fashion |
| **creative** | Purple & crimson | Art, music, entertainment |
| **minimal** | Steel grey-blue | CVs, consulting, professional |

Select a palette and click **Get Palette** to see the exact hex colour codes used for primary, secondary, accent, background, and text layers.

---
### 🧠 Tab 4 — AI UX Reviewer

After generating or loading a page, click **🔍 Run UX Review** for an instant quality audit.

The reviewer scores your page across **4 dimensions** (each out of 10):

| Dimension | What It Checks |
|---|---|
| 🗂️ **Section Coverage** | Hero, features, about, portfolio, contact, footer, CTA button |
| ♿ **Accessibility** | Alt text on images, form labels, semantic HTML tags (h1–h3), ARIA hints |
| 🎨 **Design Balance** | Fonts loaded, CSS variables used, responsive grid/flex, card components |
| 📱 **Responsiveness** | Viewport meta tag, media queries, fluid widths, mobile-friendly layout |

The overall score (0–10) is colour-coded:
- 🟢 **8–10** — Excellent
- 🟡 **6–7.9** — Good
- 🟠 **4–5.9** — Needs Work
- 🔴 **0–3.9** — Poor

A **Top Suggestions** panel lists up to 5 specific improvements you can make.

---
### 📋 Tab 5 — Templates

Load any of the 9 pre-built HTML templates with a single click.

#### Classic Templates *(AI-generated)*
These use the Generate Page engine, so they adapt to the title you've set:

| Template | What You Get |
|---|---|
| 🗂️ **Portfolio Website** | Hero banner, 6-card portfolio grid, about section, testimonials, contact form, footer |
| 📰 **Blog / Magazine** | Editorial hero, article grid, newsletter sign-up strip, archive sidebar |
| 🎤 **Event / Conference** | Countdown timer, speaker cards, session schedule, ticket CTA, sponsors strip |

#### ✨ Premium Templates *(hand-crafted HTML)*
These are fully custom-designed pages with unique fonts and visual identities:

| Template | Typography | Colour Scheme | Ideal Use |
|---|---|---|---|
| 📷 **Photography Studio** | Cormorant Garamond (serif) | Warm ivory & terracotta | Photographers, visual artists |
| 🍽️ **Restaurant** | Playfair Display (serif) | Dark amber & gold | Restaurants, cafés, food brands |
| 🛍️ **E-commerce Store** | Nunito (rounded sans) | Coral red & white | Online shops, product launches |
| 🎨 **Creative Agency** | Syne (display) | Warm tan with marquee | Design studios, freelancers |
| 👤 **Personal CV** | IBM Plex (mono/sans) | Soft purple & slate | Resumes, personal branding |
| 💪 **Fitness Studio** | Barlow Condensed | Dark + neon green | Gyms, coaches, wellness brands |

> 💡 **How templates work:** Clicking a template button loads the full HTML into the **Generate Page → Output** code box. Then switch to **👁️ Live Preview** and click **Load Preview** to render it.

---





## 🔑 Section Keywords Reference

Use these keywords in your **Page Description** field to trigger specific sections:

| Keyword(s) | Section Generated |
|---|---|
| `hero`, `banner`, `welcome` | Full-width hero with headline & CTA button |
| `features`, `services`, `benefits` | 3-column feature grid with icons |
| `about`, `team`, `story` | About section with image & text |
| `portfolio`, `gallery`, `work` | 3–6 card portfolio grid |
| `testimonials`, `reviews`, `clients` | Testimonial cards with star ratings |
| `pricing`, `plans`, `packages` | 3-tier pricing comparison table |
| `contact`, `form`, `get in touch` | Contact form with name/email/message |
| `footer` | Footer with links & copyright |
| `faq` | Accordion-style FAQ section |
| `stats`, `numbers` | Animated counter metrics row |

---

## 💡 Pro Tips & Best Practices

- **Combine multiple sections** in one description for richer pages — the AI handles all layout automatically.
- **Use Premium templates as a starting point** — load one, then copy the HTML into a text editor to customise colours and content.
- **Run the UX Reviewer** before publishing — it catches common accessibility and responsiveness issues.
- **For best results with the Portfolio template**, upload 7 images: 1 hero + 6 portfolio cards.
- **The HTML output is fully self-contained** — no frameworks, no CDN dependencies (except Google Fonts). Just save as `.html` and open in any browser.
- **To use your page live**, copy the generated HTML code and host it on GitHub Pages, Netlify, or any static hosting provider for free.

""")



print("\n✅ HTML Quick-Styler v3 — Ocean Blue Theme, 9 Templates Ready!\n")
demo.launch(share=True,show_error=True,show_api=False)
```
